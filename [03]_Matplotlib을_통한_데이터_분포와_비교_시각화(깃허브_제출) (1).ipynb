- Îã®Ïàú ÏàòÏπòÎßåÏúºÎ°úÎäî Îç∞Ïù¥ÌÑ∞Ïùò **Ìå®ÌÑ¥**ÏùÑ Ïù¥Ìï¥ÌïòÍ∏∞ Ïñ¥Î†§ÏõÄ  
- Î∂ÑÌè¨Î•º ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú Î≥¥Ïó¨Ï£ºÎäî ÏãúÍ∞ÅÌôî Í∏∞Î≤ïÏù¥ ÌïÑÏöîÌï®  
- ÎåÄÌëúÏ†ÅÏù∏ Î∞©Î≤ï:  
  - **ÏÇ∞Ï†êÎèÑ(Scatter Plot)**: Îëê Î≥ÄÏàò Í∞ÑÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÏôÄ Î∂ÑÌè¨  
  - **ÎßâÎåÄ Í∑∏ÎûòÌîÑ(Bar Chart)**: Î≤îÏ£ºÌòï Îç∞Ïù¥ÌÑ∞Ïùò ÎπÑÍµê

# ÏÇ∞Ï†êÎèÑÏùò Í∞úÎÖêÍ≥º ÌôúÏö©

- `plt.scatter(x, y)` ‚Üí Îëê Î≥ÄÏàòÏùò Í¥ÄÍ≥ÑÎ•º Ï†êÏúºÎ°ú ÌëúÌòÑ  
- Ï£ºÏöî ÏòµÏÖò:  
  - `s`: Ï†êÏùò **ÌÅ¨Í∏∞** (Ïòà: 0~30 Î≤îÏúÑÏùò ÎûúÎç§Í∞í ‚Üí ÌÅ¨Í∏∞Í∞Ä Îã§Î¶Ñ)  
  - `c`: Ï†êÏùò **ÏÉâÏÉÅ**  
  - `alpha`: **Ìà¨Î™ÖÎèÑ** (0 ~ 1)  
  - `marker`: Ï†êÏùò **Î™®Ïñë** (`'o'`, `'s'`, `'D'`, `'*'` Îì±)  

üëâ Îã§ÏñëÌïú Î™®Ïñë¬∑ÌÅ¨Í∏∞¬∑ÏÉâÏÉÅÏùÑ Ï°∞Ìï©ÌïòÎ©¥ Îç∞Ïù¥ÌÑ∞Ïùò ÌäπÏßïÍ≥º Ìå®ÌÑ¥ÏùÑ Î™ÖÌôïÌûà ÌôïÏù∏Ìï† Ïàò ÏûàÏùå

```python
import numpy as np
import matplotlib.pyplot as plt

N = 30                           # Ï†ÑÏ≤¥ Ï†ê Í∞úÏàò
x = np.random.rand(N)            # ÏûÑÏùòÏùò x Ï¢åÌëú
y = np.random.rand(N)            # ÏûÑÏùòÏùò y Ï¢åÌëú
colors = np.random.rand(N)       # ÏÉâÏÉÅ ÎûúÎç§ ÏßÄÏ†ï
area = (30 * np.random.rand(N))**2  # Ï†êÏùò ÌÅ¨Í∏∞ (Î©¥Ï†Å Í∏∞Ï§Ä, 0~15 Î≤îÏúÑ)

plt.scatter(x, y, s=area, c=colors, alpha=0.5, marker='o')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatter Plot Example")
plt.show()
```

# ü§î ÏÉùÍ∞ÅÌï¥ Î≥¥Í∏∞

ÏïÑÎûò ÏΩîÎìú ÏÖÄÏóêÏÑú **ÍµµÍ≤å ÌëúÏãúÎêú Î∂ÄÎ∂Ñ**ÏùÑ ÏßÅÏ†ë Î∞îÍøîÍ∞ÄÎ©∞ Ïã§ÌñâÌï¥ Î≥¥ÏÑ∏Ïöî.  
Î≥ÄÌôîÍ∞Ä ÎààÏóê Î≥¥Ïù¥ÎèÑÎ°ù Íµ¨Ï≤¥Ï†ÅÏù∏ ÏúÑÏπòÎ•º Í∞ÄÏù¥ÎìúÌï©ÎãàÎã§.

1. **Ìà¨Î™ÖÎèÑ Î∞îÍæ∏Í∏∞**  
   - ÌòÑÏû¨ ÏΩîÎìú: `alpha=0.5`  
   - ÏàòÏ†ï: `alpha=1`  
   üëâ Ï†êÎì§Ïù¥ Í≤πÏπ† Îïå Ïñ¥ÎñªÍ≤å Î≥¥Ïù¥ÎäîÏßÄ ÎπÑÍµêÌï¥ Î≥¥ÏÑ∏Ïöî. ÎçîÎ∂àÏñ¥ Îç∞Ïù¥ÌÑ∞ Î∂ÑÌè¨ Ìï¥ÏÑùÏóê Ïñ¥Îñ§ Ïû•Ï†ê¬∑Îã®Ï†êÏù¥ ÏûàÏùÑÍπåÏöî?

2. **ÎßàÏª§ Î™®Ïñë Î∞îÍæ∏Í∏∞**  
   - ÌòÑÏû¨ ÏΩîÎìú: `marker='o'`  
   - ÏàòÏ†ï: `marker='*'` ÎòêÎäî `marker='s'`  
   üëâ Ïõê ÎåÄÏã† Î≥Ñ, ÏÇ¨Í∞ÅÌòï Îì±ÏúºÎ°ú Î∞îÍæ∏Î©¥ Í∑∏ÎûòÌîÑ ÎäêÎÇåÏù¥ Ïñ¥ÎñªÍ≤å Îã¨ÎùºÏßàÍπåÏöî?

3. **Ï†ê ÌÅ¨Í∏∞ Í≥†Ï†ïÌïòÍ∏∞**  
   - ÌòÑÏû¨ ÏΩîÎìú: `s=area`  
   - ÏàòÏ†ï: `s=100`  
   üëâ Î™®Îì† Ï†ê ÌÅ¨Í∏∞Î•º Í∞ôÍ≤å ÌïòÎ©¥, Î∂ÑÌè¨ Ìï¥ÏÑùÏóê Ïñ¥Îñ§ Ï†ïÎ≥¥Í∞Ä ÏÇ¨ÎùºÏßàÍπåÏöî?

4. **ÏÉâÏÉÅÏóê yÍ∞í Î∞òÏòÅÌïòÍ∏∞**  
   - ÌòÑÏû¨ ÏΩîÎìú: `c=colors`  
   - ÏàòÏ†ï: `c=y`  
   üëâ ÏÉâÏÉÅÏù¥ yÍ∞íÏùò ÌÅ¨Í∏∞Ïóê Îî∞Îùº Îã¨ÎùºÏßëÎãàÎã§. ÏúÑÏ™ΩÍ≥º ÏïÑÎûòÏ™Ω Ï†êÏù¥ ÏÉâÏúºÎ°ú Ïñ¥ÎñªÍ≤å Íµ¨Î∂ÑÎêòÎäîÏßÄ Í¥ÄÏ∞∞Ìï¥ Î≥¥ÏÑ∏Ïöî.

```python
# 1Î≤à Î¨∏Ï†ú ÎãµÏïà
plt.scatter(x, y, s=area, c=colors, alpha=1, marker='o')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Transparency: alpha=1")

plt.show()

# Ìï¥ÏÑù:
# - alpha=1 ‚Üí ÏôÑÏ†ÑÌûà Î∂àÌà¨Î™Ö ‚Üí Ï†êÏù¥ Í≤πÏπòÎ©¥ ÏïÑÎûò Ï†êÏù¥ Î≥¥Ïù¥ÏßÄ ÏïäÏùå
# - alpha=0.5 ‚Üí Î∞òÌà¨Î™Ö ‚Üí Í≤πÏπú Î∂ÄÎ∂ÑÎèÑ Ïñ¥Îäê Ï†ïÎèÑ ÌôïÏù∏ Í∞ÄÎä•
# - Îî∞ÎùºÏÑú alpha=1Î°ú ÏÑ§Ï†ïÌïòÎ©¥ Îç∞Ïù¥ÌÑ∞Ïùò Ï§ëÏ≤© Ï†ïÎèÑÎ•º ÌååÏïÖÌïòÍ∏∞ Ïñ¥Î†§Ïö¥ ÌòïÌÉúÎ°ú ÎÇòÌÉÄÎÇúÎã§.
```

```python
# 2Î≤à Î¨∏Ï†ú ÎãµÏïà
plt.scatter(x, y, s=area, c=colors, alpha=0.75, marker='*')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Marker Shape: '*' (Star)")

plt.show()

# Ìï¥ÏÑù:
# - marker='o' ‚Üí ÏõêÌòï (Í∏∞Î≥∏Í∞í, Í∞ÄÏû• Î¨¥ÎÇú)
# - marker='*' ‚Üí Î≥Ñ Î™®Ïñë ‚Üí Í∞ïÏ°∞ Ìö®Í≥º, ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú ÎöúÎ†∑ÌïòÍ≤å Íµ¨Î∂ÑÎê®
# - marker='s' ‚Üí ÏÇ¨Í∞ÅÌòï ‚Üí ÏùºÏ†ïÌïú Í≤©Ïûê ÎäêÎÇåÏúºÎ°ú ÌëúÌòÑÎê®
# - Ï¶â, ÎßàÏª§Ïùò ÌòïÌÉúÎ•º Î∞îÍæ∏Î©¥ Í∑∏ÎûòÌîÑÏùò Í∞ÄÎèÖÏÑ±Í≥º ÏãúÍ∞ÅÏ†Å Ïù∏ÏÉÅÏù¥ Îã¨ÎùºÏßÄÎäî Í≤ÉÏúºÎ°ú Ìï¥ÏÑùÎêúÎã§.
```

```python
# 3Î≤à Î¨∏Ï†ú ÎãµÏïà
plt.scatter(x, y, s=100, c=colors, alpha=0.75, marker='o')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Fixed Point Size (s=100)")

plt.show()

# Ìï¥ÏÑù:
# - ÏõêÎûòÎäî Ï†êÎßàÎã§ ÌÅ¨Í∏∞Í∞Ä Îã¨Îùº (s=area) Ï∂îÍ∞ÄÏ†ÅÏù∏ Î≥ÄÏàò Ï†ïÎ≥¥Î•º ÎèôÏãúÏóê Ï†úÍ≥µ
# - Î™®Îì† Ï†ê ÌÅ¨Í∏∞Î•º ÎèôÏùºÌïòÍ≤å ÌïòÎ©¥ ÏúÑÏπò Ï†ïÎ≥¥Îßå Ï†ÑÎã¨ÎêòÍ≥† Î∂ÄÍ∞ÄÏ†Å Ï†ïÎ≥¥Í∞Ä ÏÜêÏã§Îê®
# - Îî∞ÎùºÏÑú Ï†ê ÌÅ¨Í∏∞Î•º Í≥†Ï†ïÌïòÎ©¥ Îç∞Ïù¥ÌÑ∞ ÌëúÌòÑÎ†•Ïù¥ Ï∂ïÏÜåÎêòÎäî Í≤∞Í≥ºÎ°ú Ïù¥Ïñ¥ÏßÑÎã§.
```

```python
# 4Î≤à Î¨∏Ï†ú ÎãµÏïà
plt.scatter(x, y, s=area, c=y, alpha=0.75, marker='o')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Color Mapping by y-value")

plt.show()

# Ìï¥ÏÑù:
# - c=y ‚Üí yÍ∞í ÌÅ¨Í∏∞Ïóê Îî∞Îùº ÏÉâÏÉÅÏù¥ Îã¨ÎùºÏßê
# - yÍ∞Ä ÌÅ∞ Í∞í(ÏúÑÏ™Ω)ÏùÄ ÏßÑÌïú ÏÉâ / yÍ∞Ä ÏûëÏùÄ Í∞í(ÏïÑÎûòÏ™Ω)ÏùÄ ÏòÖÏùÄ ÏÉâÏúºÎ°ú ÌëúÌòÑ
# - Ï¶â, ÏúÑÏπò Ï†ïÎ≥¥ÏôÄ ÏÉâÏÉÅ Ï†ïÎ≥¥Í∞Ä Í≤∞Ìï©ÎêòÏñ¥ Îç∞Ïù¥ÌÑ∞ Ìï¥ÏÑùÏùò ÏßÅÍ¥ÄÏÑ±Í≥º Î™ÖÌôïÏÑ±Ïù¥ Ï¶ùÎåÄÎêúÎã§Í≥† Î≥º Ïàò ÏûàÎã§.
```

# ÎßâÎåÄ Í∑∏ÎûòÌîÑÏùò Í∞úÎÖêÍ≥º ÌôúÏö©

- `plt.bar(x, Í∞í)` ‚Üí Îç∞Ïù¥ÌÑ∞Í∞íÏùÑ **ÏÑ∏Î°ú ÏßÅÏÇ¨Í∞ÅÌòï ÎßâÎåÄ**Î°ú ÌëúÌòÑ  
- `plt.xticks(x, ÎùºÎ≤®)` ‚Üí xÏ∂ï ÎààÍ∏àÏùÑ Ïà´Ïûê ÎåÄÏã† Î≤îÏ£º(Ïó∞ÎèÑ, Ïù¥Î¶Ñ Îì±)Î°ú ÌëúÏãú  

üëâ ÎßâÎåÄ Í∑∏ÎûòÌîÑÎäî **Î≤îÏ£ºÌòï Îç∞Ïù¥ÌÑ∞ ÎπÑÍµê**ÎÇò **ÏãúÍ≥ÑÏó¥ Îç∞Ïù¥ÌÑ∞ Î≥ÄÌôî**Î•º ÌëúÌòÑÌïòÎäî Îç∞ Ïú†Ïö©Ìï®

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(3)
years = ['2010', '2011', '2012']
domestic = [6801, 7695, 8010]
foreign = [777, 1046, 1681]

plt.bar(x, domestic)   # Íµ≠ÎÇ¥ÏÑ† ÏäπÍ∞ù Ïàò
plt.bar(x, foreign)    # Íµ≠Ïô∏ÏÑ† ÏäπÍ∞ù Ïàò (Í≤πÏ≥êÏÑú ÌëúÏãúÎê®)
plt.xticks(x, years)
plt.title("Domestic vs. Foreign Passengers")
plt.show()
```

# ÎßâÎåÄ Í∑∏ÎûòÌîÑÏùò ÏÑ∏Î∂Ä Í∏∞Îä•

- `width`: ÎßâÎåÄÏùò ÎëêÍªò (Í∏∞Î≥∏Í∞í=0.8)  
  ‚Üí Í∞íÏù¥ ÏûëÏùÑÏàòÎ°ù ÎßâÎåÄÍ∞Ä ÏñáÏïÑÏßê  
- `x Í∞í Î≥¥Ï†ï`: ÎßâÎåÄ ÏúÑÏπòÎ•º ÏÇ¥Ïßù Ïù¥ÎèôÌï¥ ÏÑúÎ°ú Í≤πÏπòÏßÄ ÏïäÍ≤å Ï°∞Ï†ï  

üëâ Ïó¨Îü¨ ÏãúÍ≥ÑÏó¥ Îç∞Ïù¥ÌÑ∞Î•º ÎÇòÎûÄÌûà ÎπÑÍµêÌï† Îïå Ïú†Ïö©

```python
x = np.arange(3)
years = ['2010', '2011', '2012']
domestic = [6801, 7695, 8010]
foreign = [777, 1046, 1681]

plt.bar(x, domestic, width=0.25, label='Domestic')
plt.bar(x + 0.3, foreign, width=0.25, label='Foreign')
plt.xticks(x, years)
plt.title("Domestic vs. Foreign Passengers")
plt.legend()
plt.show()
```

# ÏãúÍ∞ÅÌôî Í∏∞Î≤ïÏùò ÏùòÏùò (Ï†ïÎ¶¨)

- **ÏÇ∞Ï†êÎèÑ** ‚Üí Î≥ÄÏàò Í∞Ñ Í¥ÄÍ≥Ñ, Î∂ÑÌè¨ Ìå®ÌÑ¥ ÌååÏïÖÏóê Ï†ÅÌï©  
- **ÎßâÎåÄ Í∑∏ÎûòÌîÑ** ‚Üí Î≤îÏ£ºÌòï Îç∞Ïù¥ÌÑ∞, Í∑∏Î£π ÎπÑÍµêÏóê Ïú†Ïö©  
- Îëê Í∑∏ÎûòÌîÑ Î™®Îëê Î∂ÑÏÑù Í≤∞Í≥ºÎ•º **ÏâΩÍ≥† ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú Ï†ÑÎã¨**ÌïòÎäî ÌïµÏã¨ ÎèÑÍµ¨  

üëâ Îç∞Ïù¥ÌÑ∞ ÏÇ¨Ïù¥Ïñ∏Ïä§ Ïã§Î¨¥ÏóêÏÑú **Î≥¥Í≥†¬∑ÏÑ§Î™Ö**Ìï† Îïå Î∞òÎìúÏãú ÌôúÏö©Îê®

# ü§î ÏÉùÍ∞ÅÌï¥ Î≥¥Í∏∞

1. **ÎßâÎåÄ ÎëêÍªò Ï°∞Ï†à**  
   - ÌòÑÏû¨ ÏΩîÎìú: `width=0.25`  
   - ÏàòÏ†ï: `width=0.5`  
   üëâ ÎßâÎåÄÍ∞Ä Îçî ÎëêÍ∫ºÏõåÏßÄÎ©¥, Îëê Í∑∏Î£πÏù¥ Í≤πÏ≥êÏÑú Î≥¥ÏùºÍπåÏöî?

2. **ÎßâÎåÄ Í∞Ñ Í∞ÑÍ≤© Î∞îÍæ∏Í∏∞**  
   - ÌòÑÏû¨ ÏΩîÎìú: `x + 0.3`  
   - ÏàòÏ†ï: `x + 0.5`  
   üëâ Îëê Í∑∏Î£π ÎßâÎåÄÍ∞Ä Îçî Î©ÄÏñ¥ÏßÄÎ©¥ ÎπÑÍµêÍ∞Ä Îçî Ïâ¨Ïö∏ÍπåÏöî, Ïñ¥Î†§Ïö∏ÍπåÏöî?

3. **ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞ Ï∂îÍ∞ÄÌïòÍ∏∞**  
   - ÏÉàÎ°úÏö¥ Î¶¨Ïä§Ìä∏: `international = [300, 500, 700]`  
   - ÏΩîÎìúÏóê `plt.bar(x - 0.3, international, width=0.25, label='International')` Ï∂îÍ∞Ä  
   üëâ ÏÑ∏ Í∑∏Î£πÏùÑ Ìïú ÌôîÎ©¥ÏóêÏÑú ÎπÑÍµêÌïòÎ©¥ Ïñ¥Îñ§ Ï∞®Ïù¥Î•º ÌôïÏù∏Ìï† Ïàò ÏûàÎÇòÏöî?

```python
# 1Î≤à Î¨∏Ï†ú ÎãµÏïà
import numpy as np
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
x = np.arange(3)   # xÏ∂ï ÏúÑÏπò: [0,1,2]
years = ['2010', '2011', '2012']
domestic = [6801, 7695, 8010]  # Íµ≠ÎÇ¥ÏÑ† ÏäπÍ∞ù Ïàò
foreign = [777, 1046, 1681]    # Íµ≠Ï†úÏÑ† ÏäπÍ∞ù Ïàò

# width=0.25 ‚Üí width=0.5 Î°ú Î≥ÄÍ≤ΩÌïòÏó¨ ÎßâÎåÄ ÎëêÍªò ÌôïÎåÄ
plt.bar(x, domestic, width=0.5, label='Domestic')
plt.bar(x + 0.5, foreign, width=0.5, label='Foreign')

plt.xticks(x, years)
plt.title("Passenger Statistics: Bar Width Adjustment")
plt.legend()
plt.show()

# Ìï¥ÏÑù:
# - ÎßâÎåÄ ÎëêÍªòÍ∞Ä ÎÑìÏñ¥ÏßÄÎ©¥ Îëê Í∑∏Î£πÏù¥ Í≤πÏ≥êÏÑú ÏãúÍ∞ÅÏ†Å Íµ¨Î∂ÑÏù¥ Ïñ¥Î†µÍ≤å ÎÇòÌÉÄÎÇúÎã§.
# - Îî∞ÎùºÏÑú widthÍ∞Ä Ïª§ÏßàÏàòÎ°ù Îç∞Ïù¥ÌÑ∞ ÎπÑÍµêÏùò Î™ÖÌôïÏÑ±Ïù¥ Ï†ÄÌïòÎêúÎã§Í≥† Ìï¥ÏÑùÎêúÎã§.
```

```python
# 2Î≤à Î¨∏Ï†ú ÎãµÏïà
import numpy as np
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
x = np.arange(3)
years = ['2010', '2011', '2012']
domestic = [6801, 7695, 8010]
foreign = [777, 1046, 1681]

# ÎßâÎåÄ Í∞ÑÍ≤©ÏùÑ Ï°∞Ï†ï: foreign ÎßâÎåÄÎ•º x+0.3 ‚Üí x+0.5 Î°ú Î≥ÄÍ≤Ω
plt.bar(x - 0.25, domestic, width=0.5, label='Domestic')
plt.bar(x + 0.5, foreign, width=0.5, label='Foreign')

plt.xticks(x, years)
plt.title("Passenger Statistics: Bar Spacing Adjustment")
plt.legend()
plt.show()

# Ìï¥ÏÑù:
# - Í∞ÑÍ≤©Ïù¥ ÎÑìÏñ¥ÏßÄÎ©¥ Îëê Í∑∏Î£π ÎßâÎåÄÍ∞Ä ÏÑúÎ°ú Î©ÄÏñ¥Ï†∏ ÎπÑÍµêÍ∞Ä ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú Ïñ¥Î†§ÏõåÏßÑÎã§.
# - Î∞òÎåÄÎ°ú Ï†ÅÏ†àÌïú Í∞ÑÍ≤©ÏóêÏÑúÎäî Í∑∏Î£π Í∞Ñ Ï∞®Ïù¥Î•º ÏâΩÍ≤å ÌååÏïÖÌï† Ïàò ÏûàÎã§.
# - Ï¶â, Í∞ÑÍ≤© Ï°∞Ï†ïÏùÄ Í∞ÄÎèÖÏÑ±Í≥º Ìï¥ÏÑù Ïö©Ïù¥ÏÑ±Ïóê ÏßÅÏ†ëÏ†ÅÏù∏ ÏòÅÌñ•ÏùÑ ÎØ∏ÏπúÎã§Í≥† Î≥º Ïàò ÏûàÎã§.
```

```python
# 3Î≤à Î¨∏Ï†ú ÎãµÏïà
import numpy as np
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
x = np.arange(3)
years = ['2010', '2011', '2012']
domestic = [6801, 7695, 8010]
foreign = [777, 1046, 1681]
international = [300, 500, 700]  # Ï∂îÍ∞ÄÎêú Íµ≠Ï†úÏÑ† Ïù¥Ïô∏ Íµ¨Í∞Ñ Îç∞Ïù¥ÌÑ∞ ÏòàÏãú

# ÏÑ∏ Í∑∏Î£π(bar 3Í∞ú)ÏùÑ Ìïú ÌôîÎ©¥Ïóê ÌëúÌòÑ
plt.bar(x, domestic, width=0.2, label='Domestic')
plt.bar(x + 0.25, foreign, width=0.2, label='Foreign')
plt.bar(x - 0.25, international, width=0.2, label='International')

plt.xticks(x, years)
plt.title("Passenger Statistics: Adding International Data")
plt.legend()
plt.show()

# Ìï¥ÏÑù:
# - ÏÑ∏ Í∑∏Î£πÏùÑ ÎèôÏãúÏóê ÏãúÍ∞ÅÌôîÌïòÎ©¥ Îç∞Ïù¥ÌÑ∞ Í∞Ñ ÏÉÅÎåÄÏ†Å Ï∞®Ïù¥Í∞Ä ÎöúÎ†∑ÌïòÍ≤å ÎÇòÌÉÄÎÇúÎã§.
# - DomesticÏùÄ Ï†àÎåÄÏ†ÅÏúºÎ°ú ÌÅ∞ Í∞í, ForeignÏùÄ Ï§ëÍ∞Ñ Í∑úÎ™®, InternationalÏùÄ Í∞ÄÏû• ÏûëÏùÄ Í∞íÏúºÎ°ú Íµ¨Î∂ÑÎêúÎã§.
# - Ï¶â, Îã§Ï§ë Í∑∏Î£πÏùÑ Ìï®Íªò ÌëúÌòÑÌï®ÏúºÎ°úÏç® Îç∞Ïù¥ÌÑ∞ Í∞Ñ Í∑úÎ™® Ï∞®Ïù¥Î•º ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú Ïù¥Ìï¥Ìï† Ïàò ÏûàÍ≤å ÎêúÎã§.
```

# Îç∞Ïù¥ÌÑ∞ ÎπÑÏú®Í≥º Î∂ÑÌè¨ ÌëúÌòÑÏùò ÌïÑÏöîÏÑ±

- Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑùÏóêÏÑúÎäî Îã®Ïàú Í∞íÎøêÎßå ÏïÑÎãàÎùº **Î≤îÏ£ºÎ≥Ñ ÎπÑÏú®**Ïù¥ÎÇò **Í∞íÏùò ÌÅ¨Í∏∞ Ï∞®Ïù¥**Î•º ÌïúÎààÏóê Î≥¥Ïó¨Ï£ºÎäî Í≤ÉÏù¥ Ï§ëÏöîÌï®  
- ÎåÄÌëúÏ†Å Í∏∞Î≤ï:  
  - **ÌååÏù¥ Ï∞®Ìä∏(Pie Chart)** ‚Üí Î≤îÏ£ºÌòï Îç∞Ïù¥ÌÑ∞ ÎπÑÏú® ÌëúÌòÑÏóê Ï†ÅÌï©  
  - **ÌûàÌä∏Îßµ(Heatmap)** ‚Üí Ïó∞ÏÜçÏ†Å ÏàòÏπò Îç∞Ïù¥ÌÑ∞Ïùò Ìå®ÌÑ¥ ÌëúÌòÑÏóê Ï†ÅÌï©

# ÌååÏù¥ Ï∞®Ìä∏Ïùò Í∞úÎÖêÍ≥º ÌôúÏö©

- `plt.pie(data)` ‚Üí Ï†ÑÏ≤¥ ÎåÄÎπÑ Í∞Å Ìï≠Î™©Ïùò ÎπÑÏú®ÏùÑ ÏõêÌòïÏúºÎ°ú ÌëúÌòÑ  
- Ï£ºÏöî ÏòµÏÖò:

  - `explode`: ÌäπÏ†ï Ìï≠Î™©ÏùÑ Ï§ëÏã¨ÏóêÏÑú Î∂ÑÎ¶¨Ìï¥ Í∞ïÏ°∞

   - ÏàòÏ†ï: `cmap='cool'` ÎòêÎäî `cmap='Greys'`  
   üëâ Í∞ôÏùÄ Îç∞Ïù¥ÌÑ∞ÎùºÎèÑ ÏÉâÏÉÅÎßµÏóê Îî∞Îùº Ïñ¥Îñ§ Ï∞®Ïù¥Í∞Ä Î≥¥Ïù¥ÎÇòÏöî?

```python
# 3Î≤à Î¨∏Ï†ú ÎãµÏïà
import numpy as np

# Î¨¥ÏûëÏúÑ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± (ÌûàÌä∏Îßµ ÏòàÏãú)
matrix = np.random.rand(6, 6)

# ÏÉâÏÉÅÎßµ Î≥ÄÍ≤Ω: viridis ‚Üí cool
plt.imshow(matrix, cmap='cool')
plt.colorbar()
plt.title("Heatmap: Colormap = cool")
plt.show()

# ÏÉâÏÉÅÎßµ Î≥ÄÍ≤Ω: viridis ‚Üí Greys
plt.imshow(matrix, cmap='Greys')
plt.colorbar()
plt.title("Heatmap: Colormap = Greys")
plt.show()

# Ìï¥ÏÑù:
# - ÎèôÏùºÌïú Îç∞Ïù¥ÌÑ∞ÎùºÎèÑ ÏÉâÏÉÅÎßµÏóê Îî∞Îùº Ìå®ÌÑ¥Ïù¥ Îã§Î•¥Í≤å Ïù∏ÏãùÎêúÎã§.
# - cool ‚Üí Ï≤≠Î°ù/Î∂ÑÌôç Í≥ÑÏó¥ ÎåÄÎπÑÎ°ú ÎöúÎ†∑Ìïú Íµ¨Î∂ÑÏùÑ Ï†úÍ≥µ
# - Greys ‚Üí Î™ÖÏïî ÎåÄÎπÑÎ°ú Îã®ÏàúÌïòÍ≥† ÏßÅÍ¥ÄÏ†ÅÏù∏ ÌòïÌÉúÎ•º Ï†úÍ≥µ
# - Ï¶â, ÏÉâÏÉÅÎßµ ÏÑ†ÌÉùÏùÄ Îç∞Ïù¥ÌÑ∞Ïùò Ìå®ÌÑ¥ Ìï¥ÏÑù Î∞©ÏãùÏóê ÏßÅÏ†ëÏ†ÅÏù∏ ÏòÅÌñ•ÏùÑ ÎØ∏ÏπúÎã§.
```

# ÌûàÏä§ÌÜ†Í∑∏Îû®Ïùò Í∞úÎÖê

- Îç∞Ïù¥ÌÑ∞Î•º Ïó¨Îü¨ **Íµ¨Í∞Ñ(bin)**ÏúºÎ°ú ÎÇòÎàÑÍ≥†, Í∞Å Íµ¨Í∞ÑÏóê ÏÜçÌïòÎäî **Îç∞Ïù¥ÌÑ∞ Í∞úÏàò(ÎèÑÏàò, frequency)**Î•º ÎßâÎåÄÎ°ú ÌëúÌòÑ  
- **Í∞ÄÎ°úÏ∂ï(x-axis)**: Íµ¨Í∞Ñ  
- **ÏÑ∏Î°úÏ∂ï(y-axis)**: Ìï¥Îãπ Íµ¨Í∞ÑÏóê Ìè¨Ìï®Îêú Îç∞Ïù¥ÌÑ∞ Í∞úÏàò  

üëâ Îç∞Ïù¥ÌÑ∞Ïùò Î∂ÑÌè¨Î•º ÌïúÎààÏóê ÌååÏïÖÌï† Ïàò ÏûàÎäî **Í∏∞Î≥∏Ï†ÅÏù∏ ÌÜµÍ≥ÑÏ†Å ÏãúÍ∞ÅÌôî ÎèÑÍµ¨**

# ÏòàÏãú: ÌïôÏÉù ÌÇ§ Îç∞Ïù¥ÌÑ∞

- ÎèôÎØºÏù¥ÎÑ§ Î∞ò 30Î™Ö ÌïôÏÉùÏùò ÌÇ§ Îç∞Ïù¥ÌÑ∞Î•º ÌûàÏä§ÌÜ†Í∑∏Îû®ÏúºÎ°ú ÌëúÌòÑ  
- `plt.hist()` ‚Üí Îç∞Ïù¥ÌÑ∞ÏôÄ Íµ¨Í∞Ñ Í∞úÏàò(`bins`)Îßå ÏßÄÏ†ïÌïòÎ©¥ Î∞îÎ°ú Í∑∏Î¶¥ Ïàò ÏûàÏùå  
- `bins` Í∞íÏóê Îî∞Î•∏ Î≥ÄÌôî:  
  - `bins=6` ‚Üí Ï†ÅÏ†àÌïú Î∂ÑÌè¨ ÌôïÏù∏ Í∞ÄÎä•  
  - `bins=12` ‚Üí ÎÑàÎ¨¥ ÏÑ∏Î∂ÑÌôîÎêòÏñ¥ Ïò§ÌûàÎ†§ Ìï¥ÏÑùÏù¥ Ïñ¥Î†§ÏõÄ

```python
import numpy as np
import matplotlib.pyplot as plt

heights = np.array([
    175, 165, 164, 164, 171, 165, 160, 169, 164, 159,
    163, 167, 163, 172, 159, 160, 156, 162, 166, 162,
    158, 167, 160, 161, 156, 172, 168, 165, 165, 177
])

plt.hist(heights, bins=6)   # 6Í∞ú Íµ¨Í∞ÑÏúºÎ°ú ÎÇòÎàà ÌûàÏä§ÌÜ†Í∑∏Îû®
plt.xlabel("Height")
plt.ylabel("Frequency")
plt.title("Distribution of Student Heights")
plt.show()
```

# ÎàÑÏ†Å ÌûàÏä§ÌÜ†Í∑∏Îû® (Cumulative Histogram)

- `cumulative=True` ÏòµÏÖò ‚Üí Í∞Å Íµ¨Í∞ÑÍπåÏßÄÏùò ÎàÑÏ†Å Í∞úÏàòÎ•º ÌëúÏãú  
- Ï†ÑÏ≤¥ ÌïôÏÉù Ïàò = 30Î™Ö ‚Üí Í∞ÄÏû• Ïò§Î•∏Ï™Ω ÎßâÎåÄÍπåÏßÄÏùò Ìï©ÏùÄ 30  
- ÎàÑÏ†Å ÌûàÏä§ÌÜ†Í∑∏Îû®ÏùÄ Î∂ÑÌè¨Îøê ÏïÑÎãàÎùº **ÎàÑÏ†Å Ìï© Ï∂îÏÑ∏**ÎèÑ Ìï®Íªò Î≥¥Ïó¨Ï§å

```python
plt.hist(heights, bins=6, cumulative=False, alpha=0.6, label="Normal")
plt.hist(heights, bins=6, cumulative=True, alpha=0.6, label="Cumulative")
plt.xlabel("Height")
plt.ylabel("Frequency")
plt.legend(loc="upper left")
plt.title("Normal vs. Cumulative Histogram")
plt.show()
```

# ü§î ÏÉùÍ∞ÅÌï¥ Î≥¥Í∏∞

1. **Íµ¨Í∞Ñ Ïàò Î∞îÍæ∏Í∏∞**  
   - ÌòÑÏû¨ ÏΩîÎìú: `bins=6`  
   - ÏàòÏ†ï: `bins=12`  
   üëâ Íµ¨Í∞ÑÏùÑ ÏÑ∏Î∂ÑÌôîÌïòÎ©¥ Î∂ÑÌè¨Í∞Ä Ïñ¥ÎñªÍ≤å Îã¨Îùº Î≥¥Ïù¥ÎÇòÏöî? Î∞úÌëúÏö©/Î∂ÑÏÑùÏö© Ï§ë Ïñ¥Îäê ÏÉÅÌô©ÏóêÏÑú Ïù¥ ÌëúÌòÑÏùÑ ÏÑ†ÌÉùÌïòÍ≤†ÏäµÎãàÍπå?

3. **ÏÉâÏÉÅ ÏßÄÏ†ïÌïòÍ∏∞**  
   - ÌòÑÏû¨ ÏΩîÎìú: `plt.hist(heights, bins=6)`  
   - ÏàòÏ†ï: `plt.hist(heights, bins=6, color='orange')`  
   üëâ ÏÉâÏùÑ Î∞îÍæ∏Î©¥ Í∞ÄÎèÖÏÑ±Ïù¥ Ïñ¥ÎñªÍ≤å Îã¨ÎùºÏßÄÎÇòÏöî?

4. **ÌÇ§Ïùò ÌèâÍ∑†ÏÑ† Ï∂îÍ∞ÄÌïòÍ∏∞**  
   - `plt.axvline(heights.mean(), color='red', linestyle='dashed', linewidth=2)` Ï∂îÍ∞Ä

```python
# 3Î≤à Î¨∏Ï†ú ÎãµÏïà
plt.hist(heights, bins=6, color='orange')

plt.xlabel("Height")
plt.ylabel("Frequency")
plt.title("Histogram: Custom Color (Orange)")
plt.show()

# Ìï¥ÏÑù:
# - ÏÉâÏÉÅÏùÑ ÏßÄÏ†ïÌïòÎ©¥ Í∑∏ÎûòÌîÑÏùò Í∞ÄÎèÖÏÑ±Ïù¥ Ìñ•ÏÉÅÎêúÎã§.
# - ÌäπÌûà Î∞úÌëúÎÇò Î¶¨Ìè¨Ìä∏ÏóêÏÑúÎäî ÏÉâÏÉÅ ÎåÄÎπÑÎ•º ÌÜµÌï¥ Îç∞Ïù¥ÌÑ∞ Î©îÏãúÏßÄÍ∞Ä ÎçîÏö± ÏÑ†Î™ÖÌïòÍ≤å Ï†ÑÎã¨ÎêúÎã§.
# - Ï¶â, ÏÉâÏÉÅÏùÄ Îã®Ïàú ÎØ∏Ï†Å ÏöîÏÜåÍ∞Ä ÏïÑÎãàÎùº Îç∞Ïù¥ÌÑ∞ Ï†ÑÎã¨Î†• Ìñ•ÏÉÅ ÏöîÏù∏ÏúºÎ°ú Ìï¥ÏÑùÎêúÎã§.
```

```python
# 4Î≤à Î¨∏Ï†ú ÎãµÏïà
plt.hist(heights, bins=6, color="lightgray", edgecolor="black")

# ÌèâÍ∑†ÏÑ† Ï∂îÍ∞Ä: Îπ®Í∞Ñ Ï†êÏÑ†
plt.axvline(heights.mean(), color='red', linestyle='dashed', linewidth=2)

plt.xlabel("Height")
plt.ylabel("Frequency")
plt.title("Histogram with Mean Line")
plt.show()

# Ìï¥ÏÑù:
# - ÌèâÍ∑†ÏÑ†ÏùÑ Ï∂îÍ∞ÄÌïòÎ©¥ Îç∞Ïù¥ÌÑ∞Ïùò Ï§ëÏã¨ Í≤ΩÌñ•ÏùÑ ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú ÌôïÏù∏Ìï† Ïàò ÏûàÎã§.
# - Î∂ÑÌè¨Ïùò Ï¢åÏö∞ Í∑†ÌòïÏùÑ ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú ÌååÏïÖÌï† Ïàò ÏûàÏñ¥, Ìï¥ÏÑùÏùò Ïã†Î¢∞ÏÑ±Ïù¥ ÎÜíÏïÑÏßÑÎã§.
# - Ï¶â, ÌèâÍ∑†ÏÑ†ÏùÄ Îç∞Ïù¥ÌÑ∞Ïùò Ï§ëÏã¨Í∞íÏùÑ Í∞ïÏ°∞ÌïòÏó¨ Î∂ÑÏÑùÏ†Å ÌÜµÏ∞∞ÏùÑ Ï†úÍ≥µÌïúÎã§Í≥† Ïù¥Ìï¥Îê† Ïàò ÏûàÎã§.
```

# ÏÉÅÏûê ÏàòÏóº(Box Plot)Ïùò ÌïÑÏöîÏÑ±

- ÌèâÍ∑†ÎßåÏúºÎ°úÎäî Îç∞Ïù¥ÌÑ∞ Î∂ÑÌè¨Ïùò ÌäπÏßïÏùÑ Ï∂©Î∂ÑÌûà ÏÑ§Î™ÖÌïòÍ∏∞ Ïñ¥Î†§ÏõÄ  
- **ÏÉÅÏûê ÏàòÏóº Í∑∏Î¶º(Box Plot)**ÏùÄ  
  - ÏµúÏÜüÍ∞í, ÏµúÎåìÍ∞í  
  - Ï§ëÏïôÍ∞í (Q2, Median)  
  - ÏÇ¨Î∂ÑÏúÑÏàò(Q1, Q3)  
  - Ïù¥ÏÉÅÏπò(Outlier)  
  Î•º ÌïúÎààÏóê Î≥¥Ïó¨Ï§å  

üëâ Îç∞Ïù¥ÌÑ∞Ïùò Ìù©Ïñ¥Ïßê(Î∂ÑÏÇ∞)Í≥º Ï§ëÏïô Í≤ΩÌñ•(Ï§ëÏïôÍ∞í)ÏùÑ ÎèôÏãúÏóê ÌëúÌòÑÌïòÎäî Í∞ïÎ†•Ìïú ÏãúÍ∞ÅÌôî Î∞©Î≤ï

# Box PlotÏùò Íµ¨ÏÑ± ÏöîÏÜå

- **ÏÉÅÏûê(Box)**: Îç∞Ïù¥ÌÑ∞ Ï§ëÍ∞Ñ 50% (Q1~Q3, IQR Î≤îÏúÑ)  
  - Ï§ëÏïôÍ∞í(Q2, Median): ÏÉÅÏûê ÎÇ¥Î∂ÄÏùò ÏÑ†  
  - Q1 (ÌïòÏúÑ 25%), Q3 (ÏÉÅÏúÑ 25%)  

- **ÏàòÏóº(Whisker)**: Îç∞Ïù¥ÌÑ∞ Î∂ÑÌè¨ Î≤îÏúÑÎ•º ÎÇòÌÉÄÎÇ¥Îäî ÏÑ†  
  - ÌïòÎã® ÏàòÏóº: Q1 - 1.5 √ó IQR Ïù¥ÏÉÅ Ï§ë Í∞ÄÏû• ÏûëÏùÄ Í∞í  
  - ÏÉÅÎã® ÏàòÏóº: Q3 + 1.5 √ó IQR Ïù¥Ìïò Ï§ë Í∞ÄÏû• ÌÅ∞ Í∞í  
  - ÏàòÏóº Î∞ñ Í∞í = **Ïù¥ÏÉÅÏπò(Outlier)**, Ï†êÏúºÎ°ú ÌëúÏãú  

üëâ ÌïòÎÇòÏùò Í∑∏Î¶º ÏïàÏóêÏÑú Ïó¨Îü¨ ÏßëÎã® Îç∞Ïù¥ÌÑ∞Ïùò Î∂ÑÌè¨ ÎπÑÍµêÍ∞Ä Í∞ÄÎä•

```python
import numpy as np
import matplotlib.pyplot as plt

# Ï†ïÍ∑ú Î∂ÑÌè¨Î•º Îî∞Î•¥Îäî ÎÇúÏàò 100Í∞ú ÏÉùÏÑ±
rand_data = np.random.randn(100)

plt.boxplot(rand_data)
plt.title("Box Plot Example (Single Data)")
plt.show()
```

```python
np.random.seed(85)

data1 = np.random.normal(100, 10, 200)  # ÌèâÍ∑† 100, ÌëúÏ§ÄÌé∏Ï∞® 10
data2 = np.random.normal(100, 40, 200)  # ÌèâÍ∑† 100, ÌëúÏ§ÄÌé∏Ï∞® 40
data3 = np.random.normal(80, 40, 200)   # ÌèâÍ∑† 80, ÌëúÏ§ÄÌé∏Ï∞® 40
data4 = np.random.normal(80, 60, 200)   # ÌèâÍ∑† 80, ÌëúÏ§ÄÌé∏Ï∞® 60

plt.boxplot([data1, data2, data3, data4],
            labels=['Data1', 'Data2', 'Data3', 'Data4'])
plt.title("Box Plot Comparison (4 Groups)")
plt.show()
```

# Box PlotÏùò Ìï¥ÏÑù

- **Data1 vs. Data2**: ÌèâÍ∑†ÏùÄ Í∞ôÏßÄÎßå Data2Îäî Î∂ÑÏÇ∞(Ìù©Ïñ¥Ïßê)Ïù¥ ÌÅº  
- **Data3 vs. Data4**: ÌèâÍ∑†ÏùÄ Í∞ôÏßÄÎßå Data4Í∞Ä Îçî ÎÑìÍ≤å ÌçºÏ†∏ ÏûàÏùå  
- Ïù¥ÏÉÅÏπò(outlier)Í∞Ä Ï†êÏúºÎ°ú ÌëúÏãúÎêòÎ©¥ Îç∞Ïù¥ÌÑ∞Ïóê ÌäπÏûáÍ∞íÏù¥ ÏûàÎã§Îäî ÏùòÎØ∏  

üëâ Box PlotÏùÄ ÌèâÍ∑† Ï∞®Ïù¥Îøê ÏïÑÎãàÎùº, **Î∂ÑÏÇ∞¬∑Ï§ëÏïôÍ∞í¬∑Ïù¥ÏÉÅÏπò**ÍπåÏßÄ Ìï®Íªò Î≥¥Ïó¨Ï§å

# ü§î ÏÉùÍ∞ÅÌï¥ Î≥¥Í∏∞

1. **Îç∞Ïù¥ÌÑ∞ Ïàò Î≥ÄÍ≤ΩÌïòÍ∏∞**  
   - ÌòÑÏû¨ ÏΩîÎìú: `200`Í∞ú ÏÉòÌîå  
   - ÏàòÏ†ï: `50`Í∞ú ÏÉòÌîå (`np.random.normal(100, 10, 50)`)  
   üëâ Îç∞Ïù¥ÌÑ∞ Í∞úÏàòÍ∞Ä Ï§ÑÏñ¥Îì§Î©¥ Î∂ÑÌè¨ÏôÄ Ïù¥ÏÉÅÏπòÎäî Ïñ¥ÎñªÍ≤å Îã¨ÎùºÏßàÍπåÏöî?

2. **Î∂ÑÏÇ∞ ÌÅ¨Í∏∞ Î∞îÍæ∏Í∏∞**  
   - ÌòÑÏû¨ ÏΩîÎìú: `np.random.normal(100, 40, 200)`  
   - ÏàòÏ†ï: `np.random.normal(100, 5, 200)`  
   üëâ Î∂ÑÏÇ∞Ïù¥ Ï§ÑÏñ¥Îì§Î©¥ ÏÉÅÏûê(Box)ÏôÄ ÏàòÏóº(Whisker)Ïùò ÌÅ¨Í∏∞Í∞Ä Ïñ¥ÎñªÍ≤å Î≥ÄÌïòÎÇòÏöî?

3. **Ïù¥ÏÉÅÏπò Í¥ÄÏ∞∞ÌïòÍ∏∞**  
   - `np.random.normal(100, 10, 200)` Îç∞Ïù¥ÌÑ∞Ïóê  
     `+ [200, 250]` Í∞íÏùÑ Ï∂îÍ∞ÄÌï¥ Î≥¥ÏÑ∏Ïöî.  
   üëâ Box PlotÏóê Ïù¥ÏÉÅÏπò Ï†êÏù¥ Ïñ¥ÎñªÍ≤å ÌëúÏãúÎêòÎÇòÏöî?

4. **Í∞ÄÎ°ú Box PlotÏúºÎ°ú ÌëúÌòÑÌïòÍ∏∞**  
   - ÌòÑÏû¨ ÏΩîÎìú: `plt.boxplot(..., labels=...)`  
   - ÏàòÏ†ï: `plt.boxplot(..., labels=..., vert=False)`  
   üëâ Í∞ÄÎ°ú Î∞©Ìñ• Box PlotÏùÄ Ïñ¥Îñ§ Ïû•Ï†êÏù¥ ÏûàÏùÑÍπåÏöî?

```python
# 1Î≤à Î¨∏Ï†ú ÎãµÏïà
import numpy as np
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ Í∞úÏàòÎ•º 200 ‚Üí 50ÏúºÎ°ú Ï§ÑÏûÑ
rand_data = np.random.normal(100, 10, 50)

plt.boxplot(rand_data)
plt.title("Box Plot: Sample Size = 50")
plt.show()

# Ìï¥ÏÑù:
# - Îç∞Ïù¥ÌÑ∞ ÏàòÍ∞Ä Ï§ÑÏñ¥Îì§Î©¥ Î∂ÑÌè¨Í∞Ä Î∂àÏïàÏ†ïÌïòÍ≤å ÎÇòÌÉÄÎÇúÎã§.
# - Ïù¥ÏÉÅÏπò(outlier)ÎèÑ Îçî ÎπàÎ≤àÌïòÍ≤å Í¥ÄÏ∞∞Îê† Ïàò ÏûàÎã§.
# - Ï¶â, ÏÉòÌîå ÏàòÎäî Î∂ÑÌè¨ ÏïàÏ†ïÏÑ±Í≥º Ïù¥ÏÉÅÏπò ÌÉêÏßÄ Ïã†Î¢∞ÎèÑÏóê ÏßÅÏ†ëÏ†ÅÏúºÎ°ú ÏòÅÌñ•ÏùÑ Ï§ÄÎã§Í≥† Ìï¥ÏÑùÎêúÎã§.
```

```python
# 2Î≤à Î¨∏Ï†ú ÎãµÏïà
# Î∂ÑÏÇ∞ Ï§ÑÏù¥Í∏∞: ÌëúÏ§ÄÌé∏Ï∞® 40 ‚Üí 5
rand_data = np.random.normal(100, 5, 200)

plt.boxplot(rand_data)
plt.title("Box Plot: Variance Reduced (std=5)")
plt.show()

# Ìï¥ÏÑù:
# - Î∂ÑÏÇ∞Ïù¥ ÏûëÏïÑÏßÄÎ©¥ Îç∞Ïù¥ÌÑ∞Í∞Ä ÌèâÍ∑† Í∑ºÏ≤òÏóê Î∞ÄÏßëÌïúÎã§.
# - Îî∞ÎùºÏÑú Box(ÏÇ¨Î∂ÑÏúÑ Î≤îÏúÑ, IQR)ÏôÄ Whisker(ÏàòÏóº)Ïùò Í∏∏Ïù¥Í∞Ä ÏßßÏïÑÏßÑÎã§.
# - Ï¶â, Î∂ÑÏÇ∞ Ï∂ïÏÜåÎäî Îç∞Ïù¥ÌÑ∞Ïùò Î≥ÄÎèôÏÑ±Ïù¥ ÎÇÆÏùåÏùÑ ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú Î≥¥Ïó¨Ï§ÄÎã§Í≥† Ïù¥Ìï¥Îê† Ïàò ÏûàÎã§.
```

```python
# 3Î≤à Î¨∏Ï†ú ÎãµÏïà
# Í∏∞Î≥∏ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
rand_data = np.random.normal(100, 10, 200)

# Í∑πÎã®Ï†ÅÏù∏ Ïù¥ÏÉÅÏπò Ï∂îÍ∞Ä
rand_data2 = np.concatenate((rand_data, [200, 250]))

plt.boxplot(rand_data2)
plt.title("Box Plot: With Outliers")
plt.show()

# Ìï¥ÏÑù:
# - Box PlotÏóêÏÑúÎäî IQR(ÏÇ¨Î∂ÑÏúÑ Î≤îÏúÑ)Ïùò 1.5Î∞∞Î•º Î≤óÏñ¥ÎÇòÎäî Í∞íÏù¥ Ïù¥ÏÉÅÏπòÎ°ú ÌëúÏãúÎêúÎã§.
# - Ïó¨Í∏∞ÏÑúÎäî 200, 250Ïù¥ Î≤îÏúÑ Î∞ñÏóê ÏúÑÏπòÌïòÏó¨ ÏûëÏùÄ Ï†ê(point)ÏúºÎ°ú Îî∞Î°ú ÎÇòÌÉÄÎÇúÎã§.
# - Ï¶â, Box PlotÏùÄ Îç∞Ïù¥ÌÑ∞ Î∂ÑÌè¨ÎøêÎßå ÏïÑÎãàÎùº Ïù¥ÏÉÅÏπòÎ•º ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú ÌÉêÏßÄÌïòÎäî ÎèÑÍµ¨Î°ú Ìï¥ÏÑùÎê† Ïàò ÏûàÎã§.
```

```python
# 4Î≤à Î¨∏Ï†ú ÎãµÏïà
rand_data = np.random.normal(100, 10, 200)

plt.boxplot(rand_data, vert=False)  # Í∞ÄÎ°ú Î∞©Ìñ• ÌëúÏãú
plt.title("Horizontal Box Plot")
plt.show()

# Ìï¥ÏÑù:
# - Í∞ÄÎ°ú Î∞©Ìñ• Box PlotÏùÄ Ïó¨Îü¨ Î≤îÏ£º Îç∞Ïù¥ÌÑ∞Î•º ÎÇòÎûÄÌûà ÎπÑÍµêÌï† Îïå Ïú†Ïö©ÌïòÎã§.
# - ÌäπÌûà Î≤îÏ£ºÎ™ÖÏù¥ Í∏∏Í±∞ÎÇò ÎßéÏùÄ Í≤ΩÏö∞, Í∞ÄÎèÖÏÑ±Ïù¥ ÏàòÏßÅÎ≥¥Îã§ ÎÜíÍ≤å ÎÇòÌÉÄÎÇúÎã§.
# - Ï¶â, Îç∞Ïù¥ÌÑ∞ ÌäπÏÑ±Í≥º Î∂ÑÏÑù Î™©Ï†ÅÏóê Îî∞Îùº Í∞ÄÎ°ú/ÏÑ∏Î°ú Î∞©Ìñ•ÏùÑ ÏÑ†ÌÉùÌï† Ïàò ÏûàÎã§Í≥† Ïù¥Ìï¥ÎêúÎã§.
```

# Í∑∏ÎûòÌîÑÏùò ÌÅ¨Í∏∞ÏôÄ Í∞ÄÎèÖÏÑ±

- Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑùÏóêÏÑú Í∑∏ÎûòÌîÑÎäî Îã®ÏàúÌûà Í∑∏Î¶¨Îäî Í≤É Ïù¥ÏÉÅ ‚Üí **ÏñºÎßàÎÇò ÏâΩÍ≤å Ìï¥ÏÑùÌï† Ïàò ÏûàÎäîÍ∞Ä**Í∞Ä Ï§ëÏöî  
- Î™©Ï†ÅÏóê Îî∞Îùº Í∑∏ÎûòÌîÑ ÌÅ¨Í∏∞Î•º Ï°∞Ï†ïÌï¥Ïïº Ìï® (Î∞úÌëúÏö© / Î≥¥Í≥†ÏÑúÏö© / ÎÖºÎ¨∏Ïö©)  
- ÎòêÌïú, Í≤©Ïûê(Grid)Î•º Ï∂îÍ∞ÄÌïòÎ©¥ **Í∞íÏùò ÏúÑÏπòÏôÄ Ï∂îÏÑ∏**Î•º Ï†ïÎ∞ÄÌïòÍ≤å ÌååÏïÖ Í∞ÄÎä•

# Í∑∏ÎûòÌîÑ ÌÅ¨Í∏∞ Ï°∞Ï†à (figsize)

- `plt.figure(figsize=(Í∞ÄÎ°ú, ÏÑ∏Î°ú))` ‚Üí Í∑∏ÎûòÌîÑ ÌÅ¨Í∏∞ ÏßÅÏ†ë ÏÑ§Ï†ï  
- Îã®ÏúÑ: Ïù∏Ïπò(inch)  
- Ïòà: `figsize=(4.2, 3.6)` ‚Üí Í∞ÄÎ°ú 4.2, ÏÑ∏Î°ú 3.6  

üëâ Î∞úÌëúÏö© = ÌÅ¨Í≤å, ÎÖºÎ¨∏Ïö© = ÏûëÍ≤å, Î™©Ï†ÅÏóê ÎßûÍ≤å Ï°∞Ï†ï

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0, 2*np.pi, 200)
Y = np.sin(X)

plt.figure(figsize=(4.2, 3.6))  # ÌÅ¨Í∏∞ ÏßÄÏ†ï
plt.plot(X, Y)
plt.title("Sine Curve with Custom Size")
plt.show()
```

# Í≤©Ïûê(Grid) Ï∂îÍ∞ÄÏôÄ Ïä§ÌÉÄÏùºÎßÅ

- `plt.grid()` ‚Üí Í∏∞Î≥∏ Í≤©ÏûêÏÑ† Ï∂îÍ∞Ä  
- Ïù∏Ïûê Ï°∞Ï†à Í∞ÄÎä•:  
  - `color` ‚Üí ÏÑ† ÏÉâÏÉÅ (Ïòà: `'r'`, `'blue'`)  
  - `linestyle` ‚Üí `'-'`(Ïã§ÏÑ†), `'--'`(Ï†êÏÑ†), `':'`(Ï†ê)  
  - `linewidth` ‚Üí ÏÑ† ÎëêÍªò  
- Ï∂ï ÏÑ†ÌÉù:  
  - `axis='x'` ‚Üí ÏÑ∏Î°úÏÑ†Îßå  
  - `axis='y'` ‚Üí Í∞ÄÎ°úÏÑ†Îßå  
  - `axis='both'` ‚Üí Ï†ÑÏ≤¥ ÌëúÏãú

```python
plt.figure(figsize=(4.2, 3.6))
plt.plot(X, Y)

# Í∏∞Î≥∏ Í≤©Ïûê
plt.grid()
plt.title("Sine Curve with Default Grid")
plt.show()
```

```python
plt.figure(figsize=(4.2, 3.6))
plt.plot(X, Y)

# Ïä§ÌÉÄÏùº Ï†ÅÏö©
plt.grid(color='r', linestyle='dotted', linewidth=2, axis='both')
plt.title("Sine Curve with Custom Grid")
plt.show()
```

# ü§î ÏÉùÍ∞ÅÌï¥ Î≥¥Í∏∞

1. **Í∑∏ÎûòÌîÑ ÌÅ¨Í∏∞ ÌÇ§Ïö∞Í∏∞**  
   - ÌòÑÏû¨ ÏΩîÎìú: `figsize=(4.2, 3.6)`  
   - ÏàòÏ†ï: `figsize=(8, 6)`  
   üëâ Î∞úÌëúÏö© Í∑∏ÎûòÌîÑÎùºÎ©¥ Ïù¥Îü∞ ÌÅ¨Í∏∞Í∞Ä Ï†ÅÏ†àÌï†ÍπåÏöî? Î≥¥Í≥†ÏÑúÏö©Ïù¥ÎùºÎ©¥ Ïñ¥Îñ®ÍπåÏöî?

2. **Í≤©Ïûê Ï∂ï Ï†úÌïúÌïòÍ∏∞**  
   - ÌòÑÏû¨ ÏΩîÎìú: `axis='both'`  
   - ÏàòÏ†ï: `axis='y'`  
   üëâ yÏ∂ï Í∞í ÏùΩÍ∏∞Îäî Ïâ¨ÏõåÏßÄÍ≥†, xÏ∂ï Í∞í ÎπÑÍµêÎäî Ïñ¥ÎñªÍ≤å Îã¨ÎùºÏßÄÎÇòÏöî?

3. **Í≤©Ïûê Ïä§ÌÉÄÏùº Î∞îÍæ∏Í∏∞**  
   - ÌòÑÏû¨ ÏΩîÎìú: `linestyle='dotted', color='r'`  
   - ÏàòÏ†ï: `linestyle='--', color='gray', linewidth=1`  
   üëâ Í∑∏ÎûòÌîÑÏùò Í∞ÄÎèÖÏÑ±Ïù¥ Îçî Ï¢ãÏïÑÏßÄÎÇòÏöî?

4. **Í≥°ÏÑ†Í≥º Í≤©ÏûêÏùò Ï°∞Ìôî ÌôïÏù∏ÌïòÍ∏∞**  
   - ÏÉâÏÉÅ/Ïä§ÌÉÄÏùºÏùÑ Î∞îÍøîÍ∞ÄÎ©∞ ‚ÄúÎç∞Ïù¥ÌÑ∞ Í≥°ÏÑ†‚ÄùÍ≥º ‚ÄúÍ≤©ÏûêÏÑ†‚Äù Ï§ë Î¨¥ÏóáÏù¥ Îçî Ïûò ÎìúÎü¨ÎÇòÎäîÏßÄ ÌôïÏù∏Ìï¥ Î≥¥ÏÑ∏Ïöî.

```python
# 1Î≤à Î¨∏Ï†ú ÎãµÏïà
import numpy as np
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
X = np.linspace(0, 2*np.pi, 100)
Y = np.sin(X)

# figsize=(4.2, 3.6) ‚Üí (8, 6) ÌôïÎåÄ
plt.figure(figsize=(8, 6))
plt.plot(X, Y)

# Í≤©Ïûê Ïä§ÌÉÄÏùº
plt.grid(color='r', linestyle='dotted', linewidth=2, axis='both')
plt.title("Sine Curve: Enlarged Figure")
plt.show()

# Ìï¥ÏÑù:
# - Î∞úÌëúÏö© Í∑∏ÎûòÌîÑÏóêÏÑúÎäî (8,6)Ï≤òÎüº ÌÅ∞ ÌÅ¨Í∏∞Í∞Ä Ï†ÅÏ†àÌïòÏó¨ ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú Ïûò Ï†ÑÎã¨ÎêúÎã§.
# - Î≥¥Í≥†ÏÑúÏö©ÏùÄ (4,3) Ï†ïÎèÑÏùò ÏÜåÌòï ÌÅ¨Í∏∞Í∞Ä Ï†ÅÌï©ÌïòÏó¨ Í≥µÍ∞Ñ Ï†àÏïΩÏóê Ïú†Î¶¨ÌïòÎã§.
# - Ï¶â, Î™©Ï†Å(Î∞úÌëú vs. Î≥¥Í≥†)Ïóê Îî∞Îùº Í∑∏ÎûòÌîÑ ÌÅ¨Í∏∞ ÏÑ†ÌÉùÏù¥ Îã¨ÎùºÏ†∏Ïïº ÌïúÎã§.
```

```python
# 2Î≤à Î¨∏Ï†ú ÎãµÏïà
plt.figure(figsize=(8, 6))
plt.plot(X, Y)

# axis='both' ‚Üí axis='y' Î°ú Î≥ÄÍ≤Ω
plt.grid(color='r', linestyle='dotted', linewidth=2, axis='y')
plt.title("Sine Curve: Grid Only on y-axis")
plt.show()

# Ìï¥ÏÑù:
# - yÏ∂ïÎßå Í≤©ÏûêÍ∞Ä ÌëúÏãúÎêòÎ©¥ yÍ∞íÏùò ÏùΩÍ∏∞Îäî Îçî Ïâ¨ÏõåÏßÑÎã§.
# - Í∑∏Îü¨ÎÇò xÏ∂ï Í∞íÏùò ÎπÑÍµêÎäî Í≤©Ïûê Í∏∞Ï§ÄÏÑ†Ïù¥ ÏóÜÏñ¥ ÏÉÅÎåÄÏ†ÅÏúºÎ°ú Î∂àÌé∏ÌïòÎã§.
# - Ï¶â, Í∞ïÏ°∞Ìï† Ï∂ï(y ÎòêÎäî x)Ïóê Îî∞Îùº Í≤©Ïûê ÌëúÏãú Ï∂ïÏùÑ ÏÑ†ÌÉùÌïòÎäî Í≤ÉÏù¥ Ìö®Í≥ºÏ†ÅÏù¥Îã§.
```

```python
# 3Î≤à Î¨∏Ï†ú ÎãµÏïà
plt.figure(figsize=(8, 6))
plt.plot(X, Y)

# Ïä§ÌÉÄÏùº Î≥ÄÍ≤Ω: dotted Îπ®Í∞ï ‚Üí dashed ÌöåÏÉâ
plt.grid(linestyle='--', color='gray', linewidth=1)
plt.title("Sine Curve: Modified Grid Style")
plt.show()

# Ìï¥ÏÑù:
# - Ï†êÏÑ†(dotted)Î≥¥Îã§ dashedÍ∞Ä Îçú ÎààÏóê ÎùÑÎ©¥ÏÑúÎèÑ ÏùºÏ†ïÌïú Íµ¨Î∂ÑÏùÑ Ï†úÍ≥µÌïúÎã§.
# - ÌöåÏÉâÏùÑ ÏÇ¨Ïö©ÌïòÎ©¥ Í≥°ÏÑ†(Îç∞Ïù¥ÌÑ∞)Ïù¥ ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú Îçî Í∞ïÏ°∞ÎêúÎã§.
# - Ï¶â, Í≤©ÏûêÏùò Í∞ÄÎèÖÏÑ±Ïù¥ Ìñ•ÏÉÅÎêòÎ©∞ Îç∞Ïù¥ÌÑ∞ Í≥°ÏÑ†Í≥º Í∑†ÌòïÏùÑ Ïù¥Î£®Í≤å ÎêúÎã§.
```

```python
# 4Î≤à Î¨∏Ï†ú ÎãµÏïà
plt.figure(figsize=(8, 6))
plt.plot(X, Y, color='blue', linewidth=2)  # Îç∞Ïù¥ÌÑ∞ Í≥°ÏÑ† Í∞ïÏ°∞

# Í≤©ÏûêÏÑ†ÏùÑ ÎààÏóê Îçú ÎùÑÍ≤å ÏÑ§Ï†ï
plt.grid(linestyle=':', color='lightgray', linewidth=1)

plt.title("Sine Curve: Balance Between Curve and Grid")
plt.show()

# Ìï¥ÏÑù:
# - Îç∞Ïù¥ÌÑ∞ Í≥°ÏÑ†ÏùÑ ÏßÑÌïú ÌååÎûÄÏÉâÏúºÎ°ú ÏÑ§Ï†ïÌïòÍ≥†,
#   Í≤©ÏûêÏÑ†ÏùÑ ÏòÖÏùÄ ÌöåÏÉâÏúºÎ°ú Ï°∞Ï†ïÌïòÎ©¥ Îç∞Ïù¥ÌÑ∞ Í≥°ÏÑ†Ïù¥ ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú Îçî Í∞ïÏ°∞ÎêúÎã§.
# - Î∞òÎåÄÎ°ú Í≤©ÏûêÏÑ†ÏùÑ ÏßÑÌïòÍ≤å ÏÑ§Ï†ïÌïòÎ©¥ Îç∞Ïù¥ÌÑ∞ Í≥°ÏÑ†Î≥¥Îã§ Î∞∞Í≤ΩÏÑ†Ïù¥ Îçî ÎààÏóê ÎùÑÍ≤å ÎêúÎã§.
# - Ï¶â, Îç∞Ïù¥ÌÑ∞ÏôÄ Í≤©Ïûê Ï§ë Î¨¥ÏóáÏùÑ Í∞ïÏ°∞Ìï†ÏßÄÏóê Îî∞Îùº ÏÉâÏÉÅ¬∑Ïä§ÌÉÄÏùº Ï°∞Ìï©ÏùÑ Îã¨Î¶¨ ÏÑ†ÌÉùÌï¥Ïïº ÌïúÎã§.
```
