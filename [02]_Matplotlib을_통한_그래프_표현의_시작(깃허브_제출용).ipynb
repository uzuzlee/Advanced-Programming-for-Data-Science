# numpyÏôÄ Î∏åÎ°úÎìúÏ∫êÏä§ÌåÖ ÌôúÏö©

- `plot()` Ìï®Ïàò ÏûÖÎ†•Í∞íÏùÄ **Î¶¨Ïä§Ìä∏ÎøêÎßå ÏïÑÎãàÎùº numpy Î∞∞Ïó¥**ÎèÑ Í∞ÄÎä•Ìï®  
- numpyÏùò **Î∏åÎ°úÎìúÏ∫êÏä§ÌåÖ Í∏∞Îä•**ÏùÑ ÌôúÏö©ÌïòÎ©¥ **ÏàòÌïô Ìï®ÏàòÎ•º ÏÜêÏâΩÍ≤å ÌëúÌòÑ**Ìï† Ïàò ÏûàÏùå  

ÏòàÏãú:  
- Ìï®Ïàò: `y = x^2`  
- Î≤îÏúÑ: `x = 0 ~ 9`  
- Í≤∞Í≥º: `y Í∞íÏùÄ 0^2 ~ 9^2 ‚Üí 0 ~ 81`

```python
import numpy as np

# 0ÏóêÏÑú 9ÍπåÏßÄ Ï†ïÏàò Î∞∞Ïó¥ ÏÉùÏÑ±
x = np.arange(10)

# y = x^2 Ìï®Ïàò Í∑∏Î¶¨Í∏∞
plt.plot(x**2)
plt.show()
```

# Ï∂ï Î≤îÏúÑ ÏÑ§Ï†ïÏùò Ï§ëÏöîÏÑ±

- Ïòà: y = x¬≤ Ìï®ÏàòÎäî Í∞íÏù¥ Í∏âÍ≤©Ìûà Ïª§ÏßÄÎØÄÎ°ú yÏ∂ïÏù¥ ÏûêÎèôÏúºÎ°ú ÌÅ¨Í≤å ÏÑ§Ï†ïÎê®  

üëâ `plt.axis([xmin, xmax, ymin, ymax])` ÌòïÏãùÏúºÎ°ú ÏßÅÏ†ë Î≤îÏúÑÎ•º ÏßÄÏ†ï Í∞ÄÎä•  
- Ïòà: `plt.axis([0, 100, 0, 100])`  
- Ïù¥Î†áÍ≤å ÌïòÎ©¥ **xÏ∂ïÍ≥º yÏ∂ïÏùò Ïä§ÏºÄÏùºÏùÑ ÎèôÏùºÌïòÍ≤å ÎßûÏ∂∞** Îç∞Ïù¥ÌÑ∞ ÌäπÏÑ±ÏùÑ Îçî Î™ÖÌôïÌïòÍ≤å ÌôïÏù∏Ìï† Ïàò ÏûàÏùå  

**Í≤∞Î°†**: Ï∂ï Î≤îÏúÑ Ï°∞Ï†ïÏùÄ Îç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôîÏóêÏÑú Ï§ëÏöîÌïú Ìï¥ÏÑù ÎèÑÍµ¨ÏûÑ

```python
# Îã§Ïãú x = 0~9 Î≤îÏúÑ ÏÉùÏÑ±
x = np.arange(10)

# y = x^2 Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶º
plt.plot(x**2)

# xÏ∂ïÍ≥º yÏ∂ï Î≤îÏúÑÎ•º ÎèôÏùºÌïòÍ≤å [0, 100]ÏúºÎ°ú ÏÑ§Ï†ï
plt.axis([0, 100, 0, 100])

plt.show()
```

# Ïó¨Îü¨ Ìï®ÏàòÎ•º Ìïú ÌôîÎ©¥Ïóê Í∑∏Î¶¨Í∏∞

- Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑùÏù¥ÎÇò ÏàòÌïôÏ†Å Ìï®Ïàò ÏãúÍ∞ÅÌôîÏóêÏÑúÎäî Ïó¨Îü¨ Ìï®ÏàòÎ•º **ÎπÑÍµê**Ìï¥Ïïº ÌïòÎäî Í≤ΩÏö∞Í∞Ä ÎßéÏùå  
- Í∞ÅÍ∞Å Îî∞Î°ú Í∑∏Î¶¨Î©¥ Î≤àÍ±∞Î°≠Í≥†, ÌïúÎààÏóê ÎπÑÍµêÌïòÍ∏∞ Ïñ¥Î†§ÏõÄ  
- Îî∞ÎùºÏÑú **Ìïú ÌôîÎ©¥Ïóê ÎèôÏãúÏóê Ïó¨Îü¨ Ìï®Ïàò**Î•º ÌëúÏãúÌïòÎäî Î∞©Î≤ïÏù¥ ÌïÑÏöîÌï®  

ÏòàÏãú:  
- y‚ÇÅ = 2x  
- y‚ÇÇ = (1/3)x¬≤ + 5  
- y‚ÇÉ = -x¬≤ - 5

```python
import matplotlib.pyplot as plt
import numpy as np

# -20ÏóêÏÑú 20ÍπåÏßÄ Ï†ïÏàò ÏÉùÏÑ±
x = np.arange(-20, 20)

# ÏÑ∏ Í∞ÄÏßÄ Ìï®Ïàò Ï†ïÏùò
y1 = 2 * x
y2 = (1/3) * x**2 + 5
y3 = -x**2 - 5
```

# plot()Ïùò Ïä§ÌÉÄÏùº ÏòµÏÖò ÌôúÏö©

- plt.plot(x, y1, 'g--') ‚Üí ÎÖπÏÉâ(green), Ï†êÏÑ†(--)  
- plt.plot(x, y2, 'r^-') ‚Üí Îπ®Í∞ï(red), Ïã§ÏÑ†(-), ÏÇºÍ∞ÅÌòï(^) ÎßàÏª§  
- plt.plot(x, y3, 'b*:') ‚Üí ÌååÎûë(blue), Ï†êÏÑ†(:), Î≥Ñ(*) ÎßàÏª§  

üëâ ÏÉâÏÉÅ(Color) + ÏÑ†(Line) + ÌëúÏãù(Marker)ÏùÑ Ï°∞Ìï©Ìï¥ÏÑú  
Ìïú ÌôîÎ©¥Ïóê Ïó¨Îü¨ Ìï®ÏàòÎ•º **ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú Íµ¨Î∂Ñ**Ìï† Ïàò ÏûàÏùå  

- Ï∂ï Î≤îÏúÑÎèÑ `plt.axis([xmin, xmax, ymin, ymax])`Î°ú ÏßÄÏ†ï Í∞ÄÎä•

```python
# ÏÑ∏ Ìï®Ïàò ÎèôÏãúÏóê Í∑∏Î¶¨Í∏∞ + Ïä§ÌÉÄÏùº Ï†ÅÏö©
plt.plot(x, y1, 'g--',   # ÎÖπÏÉâ Ï†êÏÑ†
         x, y2, 'r^-',   # Îπ®Í∞ï Ïã§ÏÑ† + ÏÇºÍ∞ÅÌòï ÎßàÏª§
         x, y3, 'b*:')   # ÌååÎûë Ï†êÏÑ† + Î≥Ñ ÎßàÏª§

# xÏ∂ï, yÏ∂ï Î≤îÏúÑ ÏßÄÏ†ï
plt.axis([-30, 30, -30, 30])
plt.show()
```

# ÏÑ† Í∑∏Î¶¨Í∏∞Ïùò ÏùòÎØ∏ÏôÄ ÌôïÏû•ÏÑ±

- plot()ÏùÄ Îã®ÏàúÌûà ÏÑ† ÌïòÎÇòÎßå Í∑∏Î¶¨Îäî ÎèÑÍµ¨Í∞Ä ÏïÑÎãò  
- Ïó¨Îü¨ Ìï®ÏàòÎ•º **Ìïú ÌôîÎ©¥Ïóê ÏãúÍ∞ÅÌôî**ÌïòÎ©¥ ÎπÑÍµêÏôÄ Î∂ÑÏÑùÏù¥ Ìõ®Ïî¨ ÏßÅÍ¥ÄÏ†ÅÏûÑ  
- ÏÉâ, ÏÑ† Î™®Ïñë, ÎßàÏª§Î•º Ï†ÅÏ†àÌûà Ï°∞Ìï©ÌïòÎ©¥ ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú Î™ÖÌôïÌûà Íµ¨Î∂Ñ Í∞ÄÎä•  

üëâ Îç∞Ïù¥ÌÑ∞ ÏÇ¨Ïù¥Ïñ∏Ïä§ Ïã§Î¨¥ÏóêÏÑúÎèÑ ÎèôÏùºÌïú ÏõêÎ¶¨ Ï†ÅÏö©Îê®  
- Ïòà: Ïó¨Îü¨ Î™®Îç∏Ïùò **ÏòàÏ∏° Í≤∞Í≥º ÎπÑÍµê**, **Îç∞Ïù¥ÌÑ∞ Î∂ÑÌè¨ ÎπÑÍµê**, **Ï∂îÏÑ∏ ÎπÑÍµê**  

**Í≤∞Î°†**: plot() Ìï®ÏàòÎäî  
- Îã®Ïàú ÏÑ† Í∑∏Î¶¨Í∏∞ ÎèÑÍµ¨Í∞Ä ÏïÑÎãàÎùº,  
- Îç∞Ïù¥ÌÑ∞ ÎπÑÍµêÏôÄ Î∂ÑÏÑùÏùÑ ÏúÑÌïú **Í∞ïÎ†•Ìïú ÏãúÍ∞ÅÌôî ÏàòÎã®**

# ‚úîÔ∏è ÏùëÏö©Ìï¥ Î≥¥Í∏∞
1. Í∞ÅÍ∞Å Îã§Î•∏ ÏÉâÏÉÅÍ≥º ÎßàÏª§Î•º Ï†ÅÏö©Ìï¥ÏÑú ÎπÑÍµêÌïòÍ∏∞  
2. plt.legend()Î•º ÌôúÏö©Ìï¥ Î≤îÎ°Ä(legend)Î•º Ï∂îÍ∞ÄÌï¥ Î≥¥Í∏∞

```python
# 1. ÏÉâÏÉÅ(Color) ÏòµÏÖò
# 'r' : Îπ®Í∞ï (red)
# 'g' : Ï¥àÎ°ù (green)
# 'b' : ÌååÎûë (blue)
# 'k' : Í≤ÄÏ†ï (black)
# 'm' : ÏûêÌôç (magenta)
# 'c' : Ï≤≠Î°ù (cyan)
# 'y' : ÎÖ∏Îûë (yellow)

# 2. ÏÑ† Ïä§ÌÉÄÏùº(Line Style) ÏòµÏÖò
# '-'  : Ïã§ÏÑ† (solid line)
# '--' : Ï†êÏÑ† (dashed line)
# ':'  : Ï†ê(dot)ÏÑ† (dotted line)
# '-.' : ÏùºÏ†êÏáÑÏÑ† (dash-dot line)

# 3. ÎßàÏª§(Marker) ÏòµÏÖò
# 'o' : Ïõê(circle)
# '^' : ÏÇºÍ∞ÅÌòï(triangle up)
# 's' : ÏÇ¨Í∞ÅÌòï(square)
# '*' : Î≥Ñ(star)
# '+' : ÎçîÌïòÍ∏∞(plus)
# 'x' : Í≥±ÌïòÍ∏∞(x mark)
# 'd' : Îã§Ïù¥ÏïÑÎ™¨Îìú(diamond)

# 4. Ï°∞Ìï© ÏòàÏãú (ÏÉâÏÉÅ + ÏÑ†Ïä§ÌÉÄÏùº + ÎßàÏª§)
# plt.plot(x, y, 'Ï°∞Ìï©', label="Î≤îÎ°ÄÏóê ÌëúÏãúÎê† Ïù¥Î¶Ñ")

import matplotlib.pyplot as plt
import numpy as np

# -20ÏóêÏÑú 20ÍπåÏßÄ Ï†ïÏàò ÏÉùÏÑ±
x = np.arange(-20, 20)

# ÏÑ∏ Í∞ÄÏßÄ Ìï®Ïàò Ï†ïÏùò
y1 = 2 * x
y2 = (1/3) * x**2 + 5
y3 = -x**2 - 5

# ÏÑ∏ Ìï®Ïàò ÎèôÏãúÏóê Í∑∏Î¶¨Í∏∞ + Ïä§ÌÉÄÏùº Ï†ÅÏö©
# plt.plot(x, y, 'Ïä§ÌÉÄÏùº', label="Ïù¥Î¶Ñ")
# 'Ïä§ÌÉÄÏùº'ÏóêÎäî ÏÉâÏÉÅ(color) + ÏÑ†(line style) + ÎßàÏª§(marker)Î•º Ï°∞Ìï© Í∞ÄÎä•
# labelÏùÄ Î≤îÎ°Ä(legend)ÏóêÏÑú ÏÇ¨Ïö©Ìï† Ïù¥Î¶Ñ

plt.plot(x, y1, 'r--', label="y1")   # Îπ®Í∞ï(red) + Ï†êÏÑ†(dashed) ‚Üí y1 Í≥°ÏÑ†
plt.plot(x, y2, 'g^-', label="y2")   # Ï¥àÎ°ù(green) + Ïã§ÏÑ†(solid) + ^(ÏÇºÍ∞ÅÌòï ÎßàÏª§) ‚Üí y2 Í≥°ÏÑ†
plt.plot(x, y3, 'b*:', label="y3")   # ÌååÎûë(blue) + Ï†êÏÑ†(dotted) + *(Î≥Ñ ÎßàÏª§) ‚Üí y3 Í≥°ÏÑ†

# legendÎ•º ÌëúÏãú ‚Üí ÏúÑÏóêÏÑú ÏßÄÏ†ïÌïú labelÎì§ÏùÑ Î≤îÎ°Ä Î∞ïÏä§Î°ú ÎÇòÌÉÄÎÉÑ
plt.legend()

# xÏ∂ï, yÏ∂ï Î≤îÏúÑ ÏßÄÏ†ï (Ï¢åÏö∞ -30~30, ÏÉÅÌïò -30~30ÏúºÎ°ú Î≥¥Ïù¥Í≤å Ìï®)
plt.axis([-30, 30, -30, 30])

# Í∑∏ÎûòÌîÑ Ï∂úÎ†•
plt.show()
```

# Îã§ÏñëÌïú Îç∞Ïù¥ÌÑ∞ÏôÄ ÏÑ† ÌëúÌòÑÏùò ÌïÑÏöîÏÑ±

- Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑùÏóêÏÑúÎäî Îã®Ïàú ÏßÅÏÑ†Ïù¥ ÏïÑÎãàÎùº **Î≥µÏû°Ìïú Ìå®ÌÑ¥**ÏùÑ ÏãúÍ∞ÅÌôîÌï¥Ïïº Ìï®  
- Ïòà:  
  - ÎÇúÏàò(random) Îç∞Ïù¥ÌÑ∞ ‚Üí Î∂àÍ∑úÏπôÏ†ÅÏù∏ Ìå®ÌÑ¥  
  - Ï£ºÍ∏∞ Ìï®Ïàò(sin, cos) ‚Üí Î∞òÎ≥µÏ†ÅÏù∏ ÌååÎèô  

üëâ `plt.plot()`ÏùÄ Ïù¥Îü∞ Îã§ÏñëÌïú Îç∞Ïù¥ÌÑ∞Î•º Ìïú ÌôîÎ©¥ÏóêÏÑú ÌëúÌòÑÌï† Ïàò ÏûàÏùå

```python
import matplotlib.pyplot as plt
import numpy as np

# ÎÇúÏàò Îç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôî
N = 50
x = np.arange(N)
y = np.random.random(size=N) # 0Í≥º 1 ÏÇ¨Ïù¥Ïùò Î¨¥ÏûëÏúÑ Ïã§ÏàòÎ•º ÎΩëÏïÑÎÇ¥Îäî Ìï®Ïàò

# ÎÖπÏÉâ Ï†êÏÑ†(:), ÏÇºÍ∞ÅÌòï(^) ÎßàÏª§
plt.plot(x, y, 'g^:')
plt.show()
```

# ÎÇúÏàòÏôÄ Ï£ºÍ∏∞ Ìï®Ïàò ÏãúÍ∞ÅÌôî

- `np.random()` ‚Üí Î∂àÍ∑úÏπôÌïú ÎÇúÏàò Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±  
- `np.sin()`, `np.cos()` ‚Üí Î∞òÎ≥µÏ†ÅÏù∏ ÌååÎèô(Ï£ºÍ∏∞ Ìï®Ïàò) ÏÉùÏÑ±  

üëâ Ïã§Ï†ú Îç∞Ïù¥ÌÑ∞(Ïû°Ïùå Ìè¨Ìï®)ÏôÄ Ïù¥Î°†Ï†Å Î™®Îç∏(Ìï®Ïàò)ÏùÑ ÎπÑÍµêÌïòÎäî Îç∞ ÌôúÏö© Í∞ÄÎä•

```python
# 0 ~ 2œÄ Î≤îÏúÑÏóêÏÑú 100Í∞úÏùò Í∞í ÏÉùÏÑ±
x = np.linspace(0, np.pi * 2, 100)

# sin(x): Îπ®Í∞ï Ïã§ÏÑ†, cos(x): ÌååÎûë Ï†êÏÑ†
plt.plot(x, np.sin(x), 'r-')
plt.plot(x, np.cos(x), 'b:')
plt.show()
```

# Í∑∏ÎûòÌîÑ Ï†ÄÏû•ÌïòÍ∏∞

- ÏãúÍ∞ÅÌôî Í≤∞Í≥ºÎäî ÌôîÎ©¥Ïóê ÌëúÏãúÌï† ÎøêÎßå ÏïÑÎãàÎùº ÌååÏùºÎ°ú Ï†ÄÏû• Í∞ÄÎä•  
- Ï£ºÏöî Ìï®Ïàò:  
  - `plt.figure()` ‚Üí ÏÉà figure Í∞ùÏ≤¥ ÏÉùÏÑ±  
  - `fig.savefig("ÌååÏùºÎ™Ö.png")` ‚Üí PNG ÌååÏùºÎ°ú Ï†ÄÏû•  

üëâ ÏßÄÏõê Ìè¨Îß∑: PNG, JPG, SVG, TIF, PDF Îì± 10Ïó¨ Í∞ÄÏßÄ Ïù¥ÏÉÅ  
üëâ ÏßÄÏõê Ìè¨Îß∑ ÌôïÏù∏: `fig.canvas.get_supported_filetypes()`

```python
# sin, cos Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶¨Í≥† ÌååÏùºÎ°ú Ï†ÄÏû•
x = np.linspace(0, np.pi * 2, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), 'r-')
plt.plot(x, np.cos(x), 'b:')
fig.savefig('sin_cos_fig.png')  # PNG ÌååÏùº Ï†ÄÏû•
```

# plt.show()ÏóêÏÑú DPI Ï†ÅÏö©ÌïòÍ∏∞

- Î≥¥ÌÜµ `plt.show()`Îäî ÎîîÌè¥Ìä∏ Ìï¥ÏÉÅÎèÑ(dpi=100) Ï†ïÎèÑÎ°ú Ï∂úÎ†•Îê®  
- Îçî ÏÑ†Î™ÖÌïú Í∑∏ÎûòÌîÑÎ•º Î≥¥Í≥† Ïã∂Îã§Î©¥ **Ï∂úÎ†• Ìï¥ÏÉÅÎèÑ ÏûêÏ≤¥Î•º Ï°∞Ï†ï**Ìï† Ïàò ÏûàÏùå  

Î∞©Î≤ï:  
1. `plt.figure(dpi=Í∞í)` ‚Üí Í∞úÎ≥Ñ Í∑∏ÎûòÌîÑÎßàÎã§ ÏÑ§Ï†ï  
2. `plt.rcParams["figure.dpi"] = Í∞í` ‚Üí Ï†ÑÏ≤¥ Í∏∞Î≥∏Í∞í Î≥ÄÍ≤Ω  

üëâ ColabÏóêÏÑú Í∑∏ÎûòÌîÑÎ•º Î≥º Îïå,  
- `dpi=72` ‚Üí ÌùêÎ¶øÌïòÍ≤å Î≥¥ÏûÑ  
- `dpi=150` Ïù¥ÏÉÅ ‚Üí ÏÑ†Î™ÖÌïòÍ≤å Î≥¥ÏûÑ

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, np.pi * 2, 100)

# Í∞úÎ≥Ñ Figure Ìï¥ÏÉÅÎèÑ ÏÑ§Ï†ï (dpi=150)
plt.figure(dpi=150)
plt.plot(x, np.sin(x), 'r-', label='sin')
plt.plot(x, np.cos(x), 'b:', label='cos')
plt.legend()
plt.title("Sin and Cos Curve (dpi=150)")
plt.show()
```

```python
# Ï†ÑÏ≤¥ Ï∂úÎ†• Í∏∞Î≥∏Í∞íÏùÑ Í≥†Ìï¥ÏÉÅÎèÑÎ°ú Î≥ÄÍ≤Ω (Ïòà: 200 dpi)
plt.rcParams["figure.dpi"] = 200

x = np.linspace(0, np.pi * 2, 100)

plt.plot(x, np.sin(x), 'r-', label='sin')
plt.plot(x, np.cos(x), 'b:', label='cos')
plt.legend()
plt.title("Sin and Cos Curve (Global dpi=200)")
plt.show()
```

ÌôîÎ©¥ Ï∂úÎ†• Ïãú Í∑∏ÎûòÌîÑ ÏÑ†Î™ÖÎèÑÏùò Ï∞®Ïù¥Î•º ÎààÏúºÎ°ú ÌôïÏù∏Ìï† Ïàò ÏûàÏùå

# ‚úîÔ∏è ÏùëÏö©Ìï¥ Î≥¥Í∏∞
1. np.random()ÏúºÎ°ú 100Í∞ú ÎÇúÏàòÎ•º ÎßåÎì§Ïñ¥ ÏãúÍ∞ÅÌôîÌï¥ Î≥¥Í∏∞  
2. sin(x)ÏôÄ cos(x)Î•º Í∞ôÏùÄ ÌôîÎ©¥Ïóê Í∑∏Î¶¨Îêò, x Î≤îÏúÑÎ•º 0~4œÄÎ°ú ÌôïÏû•Ìï¥ Î≥¥Í∏∞  
3. dpi(ÌôîÏßà ÏÑ§Ï†ï)Ïóê Îî∞Îùº Ï∂úÎ†• Í≤∞Í≥ºÍ∞Ä Ïñ¥ÎñªÍ≤å Îã¨ÎùºÏßÄÎäîÏßÄ ÎπÑÍµêÌï¥ Î≥¥Í∏∞

```python
# 1Î≤à Î¨∏Ï†ú Î≥ÄÌòï ÎãµÏïà: np.random.rand(10) ‚Üí 0Í≥º 1 ÏÇ¨Ïù¥Ïùò ÎÇúÏàòÎ•º 10Í∞ú ÏÉùÏÑ±
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(10)                 # x Í∞íÏùÄ 0Î∂ÄÌÑ∞ 9ÍπåÏßÄ
y = np.random.rand(10)            # y Í∞íÏùÄ 0Í≥º 1 ÏÇ¨Ïù¥ Î¨¥ÏûëÏúÑ Ïã§Ïàò 10Í∞ú

# Ï¥àÎ°ùÏÉâ(green) + Ï†êÏÑ†(:) + ÏÇºÍ∞ÅÌòï(^) ÎßàÏª§ Ïä§ÌÉÄÏùºÎ°ú ÏãúÍ∞ÅÌôî
plt.plot(x, y, 'g^:')
plt.show()
```

```python
# 2Î≤à Î¨∏Ï†ú ÎãµÏïà: sin(x)ÏôÄ cos(x)Î•º Í∞ôÏùÄ ÌôîÎ©¥Ïóê Í∑∏Î¶¨Îêò, x Î≤îÏúÑÎ•º 0 ~ 4œÄÎ°ú ÌôïÏû•ÌïòÍ∏∞
import numpy as np
import matplotlib.pyplot as plt

# x Í∞í ÏÉùÏÑ±: 0ÏóêÏÑú 4œÄÍπåÏßÄ Íµ¨Í∞ÑÏùÑ 100Í∞úÎ°ú ÎÇòÎàî
x = np.linspace(0, np.pi * 4, 100)

# sin(x) Í∑∏ÎûòÌîÑ ‚Üí Îπ®Í∞ï Ïã§ÏÑ†
plt.plot(x, np.sin(x), 'r-', label='sin(x)')

# cos(x) Í∑∏ÎûòÌîÑ ‚Üí ÌååÎûë Ï†êÏÑ†
plt.plot(x, np.cos(x), 'b:', label='cos(x)')

# Î≤îÎ°Ä Ï∂îÍ∞Ä (sin / cos Íµ¨Î∂Ñ)
plt.legend()

# Í∑∏ÎûòÌîÑÎ•º ÌååÏùºÎ°ú Ï†ÄÏû• (ÏõêÌïòÎ©¥ Ï£ºÏÑù Ìï¥Ï†úÌï¥ÏÑú ÏÇ¨Ïö© Í∞ÄÎä•)
# plt.savefig("sin_cos.png", dpi=150)

# ÏµúÏ¢Ö Í∑∏ÎûòÌîÑ Ï∂úÎ†•
plt.show()
```

# Í∑∏ÎûòÌîÑ Í∞ÄÎèÖÏÑ±Ïùò ÌïÑÏöîÏÑ±

- Îã®ÏàúÌûà ÏÑ†Îßå Í∑∏Î¶∞ Í∑∏ÎûòÌîÑÎäî Îç∞Ïù¥ÌÑ∞Ïùò ÏùòÎØ∏Î•º Ïù¥Ìï¥ÌïòÍ∏∞ Ïñ¥Î†§ÏõÄ  
- Îî∞ÎùºÏÑú **Ï†úÎ™©(title)**, **Ï∂ï Î†àÏù¥Î∏î(xlabel, ylabel)**, **Î≤îÎ°Ä(legend)**Î•º Ï∂îÍ∞ÄÌï¥Ïïº Ìï®  
- Ïù¥Î†áÍ≤å ÌïòÎ©¥ Í∑∏ÎûòÌîÑÍ∞Ä Îçî **ÏßÅÍ¥ÄÏ†Å**Ïù¥Í≥† **ÏÑ§Î™ÖÎ†•** ÏûàÍ≤å Îê®

```python
import numpy as np
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
x = np.linspace(0, np.pi * 2, 100)

# Ï†úÎ™©, Ï∂ï Î†àÏù¥Î∏î, Î≤îÎ°Ä Ï†ÅÏö©
plt.title('Sin and Cos Curve')              # Í∑∏ÎûòÌîÑ Ï†úÎ™©
plt.plot(x, np.sin(x), 'r-', label='sin')   # Îπ®Í∞ï Ïã§ÏÑ† + Î≤îÎ°Ä ÎùºÎ≤®
plt.plot(x, np.cos(x), 'b:', label='cos')   # ÌååÎûë Ï†êÏÑ† + Î≤îÎ°Ä ÎùºÎ≤®
plt.xlabel('x value')                       # xÏ∂ï Î†àÏù¥Î∏î
plt.ylabel('y value')                       # yÏ∂ï Î†àÏù¥Î∏î
plt.legend()                                # Î≤îÎ°Ä Ï∂úÎ†•
plt.show()
```

# Ï†úÎ™©¬∑Î†àÏù¥Î∏î¬∑Î≤îÎ°Ä Ï∂îÍ∞ÄÌïòÍ∏∞

- `plt.title("Ï†úÎ™©")` ‚Üí Í∑∏ÎûòÌîÑÏùò Ï†úÎ™© ÏÑ§Ï†ï  
- `plt.xlabel("xÏ∂ï Ïù¥Î¶Ñ")`, `plt.ylabel("yÏ∂ï Ïù¥Î¶Ñ")` ‚Üí Ï∂ï ÏÑ§Î™Ö Ï∂îÍ∞Ä  
- `plt.plot(..., label="ÏÑ§Î™Ö")` ‚Üí Í∞Å Í∑∏ÎûòÌîÑÏóê Î≤îÎ°Ä Ïù¥Î¶Ñ ÏßÄÏ†ï  
- `plt.legend()` ‚Üí labelÏùÑ Î∂àÎü¨ÏôÄ Î≤îÎ°Ä Ï∂úÎ†•  
- `plt.legend(loc="upper right")` ‚Üí Î≤îÎ°Ä ÏúÑÏπò ÏßÄÏ†ï Í∞ÄÎä•

```python
# Î≤îÎ°Ä ÏúÑÏπòÎ•º Ïò§Î•∏Ï™Ω ÏÉÅÎã®Ïóê ÌëúÏãú
x = np.linspace(0, np.pi * 2, 100)

plt.title('Sin and Cos Curve')
plt.plot(x, np.sin(x), 'r-', label='sin')
plt.plot(x, np.cos(x), 'b:', label='cos')
plt.xlabel('x value')
plt.ylabel('y value')
plt.legend(loc='upper right')  # Î≤îÎ°Ä ÏúÑÏπò ÏßÄÏ†ï
plt.show()
```

```python
# 3Î≤à Î¨∏Ï†ú ÎãµÏïà: Ï†úÎ™©ÏùÑ "Sine vs Cosine"ÏúºÎ°ú Î∞îÍæ∏Í≥†, Î≤îÎ°Ä ÏúÑÏπòÎ•º "lower left"(ÏôºÏ™Ω ÏïÑÎûò)Î°ú Î≥ÄÍ≤ΩÌïòÍ∏∞

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, np.pi * 2, 100)   # 0 ~ 2œÄ Íµ¨Í∞ÑÏùÑ 100Í∞úÎ°ú ÎÇòÎàî

plt.title('Sine vs Cosine')          # 1Î≤à ÎãµÏïà - Í∑∏ÎûòÌîÑ Ï†úÎ™© ÏÑ§Ï†ï
plt.plot(x, np.sin(x), 'r-', label='sin')  # sin(x) Îπ®Í∞ï Ïã§ÏÑ†
plt.plot(x, np.cos(x), 'b:', label='cos')  # cos(x) ÌååÎûë Ï†êÏÑ†
plt.xlabel('x value')                # xÏ∂ï ÎùºÎ≤®
plt.ylabel('y value')                # yÏ∂ï ÎùºÎ≤®

plt.legend(loc='lower left')         # 2Î≤à ÎãµÏïà - Î≤îÎ°Ä ÏúÑÏπòÎ•º ÏôºÏ™Ω ÏïÑÎûòÎ°ú ÏßÄÏ†ï
plt.show()
```

# Îã®Ïùº plot() Ìï®ÏàòÏùò ÌïúÍ≥Ñ

- `plt.plot()`ÏùÄ Í∞ÑÎã®Ìïú Í∑∏ÎûòÌîÑÎ•º Îπ†Î•¥Í≤å Í∑∏Î¶¥ Îïå Ïú†Ïö©  
- ÌïòÏßÄÎßå Ïó¨Îü¨ Ï¢ÖÎ•òÏùò Í∑∏ÎûòÌîÑÎ•º ÎèôÏãúÏóê Î∞∞ÏπòÌïòÍ∏∞ÏóêÎäî ÌïúÍ≥Ñ Ï°¥Ïû¨  
- Îî∞ÎùºÏÑú **Ïó¨Îü¨ Í∞úÏùò Í∑∏ÎûòÌîÑÎ•º Ìïú ÌôîÎ©¥Ïóê Î∞∞Ïπò**Ìï† Ïàò ÏûàÎäî Î∞©Î≤ïÏù¥ ÌïÑÏöî

# subplots() Ìï®ÏàòÏùò Îì±Ïû•

- `plt.subplots(nrows, ncols)` ‚Üí Ìñâ √ó Ïó¥ Íµ¨Ï°∞Î°ú Í∑∏ÎûòÌîÑ Î∞∞Ïπò  
- Î∞òÌôò Í∞í: **Figure Í∞ùÏ≤¥ + Axes Í∞ùÏ≤¥**  
  - **Figure**: Ï†ÑÏ≤¥ Í∑∏ÎûòÌîÑ ÏòÅÏó≠ (ÌÅ∞ ÌãÄ, Ïª®ÌÖåÏù¥ÎÑà)  
  - **Axes**: Ïã§Ï†ú Í∑∏ÎûòÌîÑÍ∞Ä Í∑∏Î†§ÏßÄÎäî ÏûëÏùÄ Î∞ïÏä§  

üëâ Ïó¨Îü¨ AxesÎäî Î∞∞Ïó¥Î°ú Í¥ÄÎ¶¨ÎêòÎ©∞, `ax[0,0]`, `ax[1,1]`Ï≤òÎüº Ïù∏Îç±Ïä§Î°ú Ï†ëÍ∑º Í∞ÄÎä•

```python
import matplotlib.pyplot as plt
import numpy as np

# 2Ìñâ 2Ïó¥ subplot ÏÉùÏÑ±
fig, ax = plt.subplots(2, 2)

# ÏÇ∞Ï†êÎèÑ
X = np.random.randn(100)
Y = np.random.randn(100)
ax[0, 0].scatter(X, Y)

# ÎßâÎåÄ Í∑∏ÎûòÌîÑ
X = np.arange(10)
Y = np.random.uniform(1, 10, 10)
ax[0, 1].bar(X, Y)

# ÏÑ† Í∑∏ÎûòÌîÑ
X = np.linspace(0, 10, 100)
Y = np.cos(X)
ax[1, 0].plot(X, Y)

# 2D Ïù¥ÎØ∏ÏßÄ
Z = np.random.uniform(0, 1, (5, 5))
ax[1, 1].imshow(Z)

plt.show()
```

# AxesÏôÄ AxisÏùò Íµ¨Î∂Ñ

- **Axes** (Î≥µÏàòÌòï): Í∑∏ÎûòÌîÑÎ•º Îã¥Îäî ÏòÅÏó≠(ÌîÑÎ†àÏûÑ)  
- **Axis** (Îã®ÏàòÌòï): Ïã§Ï†ú Ï∂ï(xÏ∂ï, yÏ∂ï)  
- Ïù¥Î¶ÑÏù¥ ÎπÑÏä∑Ìï¥ ÌòºÎèôÌïòÍ∏∞ ÏâΩÏßÄÎßå,  
  üëâ Axes = Í≥µÍ∞Ñ / Axis = Ï∂ï ÏúºÎ°ú Í∏∞Ïñµ

# subplot()Ïùò Í≥†Í∏â Í∏∞Îä•

- `plt.subplots()`Îäî Ìñâ(row) √ó Ïó¥(col) Íµ¨Ï°∞Î°ú Î∞∞Ïπò  
- ÌïòÏßÄÎßå Îã®Ïàú Î∞∞ÏπòÎßåÏúºÎ°úÎäî **Í∞ÑÍ≤© Ï°∞Ï†à**Ïù¥ Ïñ¥Î†§ÏõÄ  

üëâ Ìï¥Í≤∞Ï±Ö:  
- `wspace` ‚Üí ÏàòÌèâ Í∞ÑÍ≤© Ï°∞Ï†à  
- `hspace` ‚Üí ÏàòÏßÅ Í∞ÑÍ≤© Ï°∞Ï†à  

Ïòà:  
- `wspace=0.4` ‚Üí Ï¢åÏö∞ Í∞ÑÍ≤© ÎÑìÍ≤å  
- `hspace=0` ‚Üí ÏúÑÏïÑÎûò Í∑∏ÎûòÌîÑÎ•º Î∂ôÏó¨ Î∞∞Ïπò

```python
# 2Ìñâ 3Ïó¥ subplot Î∞∞Ïπò
fig, ax = plt.subplots(2, 3)

# Í∞Å ÏúÑÏπòÏóê ÌÖçÏä§Ìä∏Î°ú Ï¢åÌëú ÌëúÏãú (ÏãúÍ∞Å ÌôïÏù∏Ïö©)
for i in range(2):
    for j in range(3):
        ax[i, j].text(0.35, 0.5, str((i, j)), fontsize=11)

plt.show()
```

# GridSpecÏùÑ ÌôúÏö©Ìïú Í≥†Í∏â Î∞∞Ïπò

- `plt.GridSpec(nrows, ncols, wspace, hspace)` ‚Üí ÏÑ∏Î∞ÄÌïú Í∞ÑÍ≤© Ï†úÏñ¥ Í∞ÄÎä•  
- ÌäπÏ†ï ÏòÅÏó≠ÏùÑ Î≥ëÌï©Ìï¥ÏÑú Îçî ÌÅ∞ Í∑∏ÎûòÌîÑÎèÑ Î∞∞Ïπò Í∞ÄÎä•  

üëâ Îã§ÏñëÌïú Îç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôîÎ•º **ÎåÄÏãúÎ≥¥Îìú ÌòïÌÉú**Î°ú Íµ¨ÏÑ± Í∞ÄÎä•

```python
# GridSpecÏúºÎ°ú Ïó¨Î∞± Ï°∞Ï†ï Î∞è ÏòÅÏó≠ Î≥ëÌï©
grid = plt.GridSpec(2, 3, wspace=0.3, hspace=0.4)

# ÏÇ∞Ï†êÎèÑ
X = np.random.randn(100)
Y = np.random.randn(100)
plt.subplot(grid[0, 0]).scatter(X, Y)

# ÎßâÎåÄ Í∑∏ÎûòÌîÑ (0Ìñâ 1~2Ïó¥ Ìï©Ïπ®)
X = np.arange(10)
Y = np.random.uniform(1, 10, 10)
plt.subplot(grid[0, 1:]).bar(X, Y)

# ÏÑ† Í∑∏ÎûòÌîÑ (1Ìñâ 0~1Ïó¥ Ìï©Ïπ®)
X = np.linspace(0, 10, 100)
Y = np.cos(X)
plt.subplot(grid[1, :2]).plot(X, Y)

# 2D Ïù¥ÎØ∏ÏßÄ
Z = np.random.uniform(0, 1, (5, 5))
plt.subplot(grid[1, 2]).imshow(Z)

plt.show()
```

# ‚úîÔ∏è ÏùëÏö©Ìï¥ Î≥¥Í∏∞
1. 3Ìñâ 2Ïó¥ Íµ¨Ï°∞Î°ú subplot Î∞∞Ïπò ÌõÑ, Í∞Å ÏúÑÏπòÏóê Îã§Î•∏ Í∑∏ÎûòÌîÑÎ•º ÎÑ£Ïñ¥Î≥¥Í∏∞  
2. wspace, hspace Í∞íÏùÑ Îã§ÏñëÌïòÍ≤å Î∞îÍøîÎ≥¥Î©∞ Í∞ÑÍ≤© Î≥ÄÌôîÎ•º ÌôïÏù∏ÌïòÍ∏∞  
3. GridSpecÏùÑ Ïù¥Ïö©Ìï¥ Ï≤´ Î≤àÏß∏ Ìñâ Ï†ÑÏ≤¥Î•º ÌïòÎÇòÏùò ÌÅ∞ Í∑∏ÎûòÌîÑÎ°ú Ìï©Ï≥êÎ≥¥Í∏∞

```python
# 3Ìñâ 2Ïó¥ subplot Î∞∞Ïπò ‚Üí Ï†ÑÏ≤¥Î•º 6Í∞úÏùò ÏûëÏùÄ ÏòÅÏó≠(axes)ÏúºÎ°ú ÎÇòÎàî
fig, ax = plt.subplots(3, 2)

# Í∞Å subplot ÏúÑÏπòÏóê (Ìñâ, Ïó¥) Ï¢åÌëú ÌÖçÏä§Ìä∏Î•º ÌëúÏãúÌïòÏó¨
# Î∞∞Ïπò Íµ¨Ï°∞(0Ìñâ0Ïó¥, 0Ìñâ1Ïó¥ ‚Ä¶)Î•º ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú ÌôïÏù∏Ìï† Ïàò ÏûàÎèÑÎ°ù Ìï®
for i in range(3):
    for j in range(2):
        ax[i, j].text(0.35, 0.5, str((i, j)), fontsize=11)

plt.show()
```

```python
# wspace, hspace Í∞ÑÍ≤© Ïã§ÌóòÏùÑ ÏúÑÌïú Ìï®Ïàò Ï†ïÏùò ‚Üí GridSpecÏùÑ ÏÇ¨Ïö©Ìï¥ÏÑú subplot(ÏÑúÎ∏åÌîåÎ°Ø) Î∞∞Ïπò Ïãú
# Í∞ÄÎ°ú Í∞ÑÍ≤©(wspace), ÏÑ∏Î°ú Í∞ÑÍ≤©(hspace)Ïù¥ Ïñ¥ÎñªÍ≤å Îã¨ÎùºÏßÄÎäîÏßÄ ÌôïÏù∏

import matplotlib.pyplot as plt
import numpy as np

def demo_wspace_hspace(wspace, hspace):
    fig = plt.figure(figsize=(6,4))  # Ï†ÑÏ≤¥ Figure ÌÅ¨Í∏∞ ÏÑ§Ï†ï
    # Ï†úÎ™©Ïóê ÌòÑÏû¨ wspace, hspace Í∞íÏùÑ ÌëúÏãú (ÎπÑÍµêÌïòÍ∏∞ ÏâΩÍ≤å)
    fig.suptitle(f"wspace={wspace}, hspace={hspace}", fontsize=12)

    # GridSpec: 2Ìñâ 3Ïó¥ Í≤©Ïûê Î∞∞Ïπò
    #   wspace ‚Üí subplot ÏÇ¨Ïù¥ Í∞ÄÎ°ú Í∞ÑÍ≤© (width space)
    #   hspace ‚Üí subplot ÏÇ¨Ïù¥ ÏÑ∏Î°ú Í∞ÑÍ≤© (height space)
    grid = plt.GridSpec(2, 3, wspace=wspace, hspace=hspace)

    # (0Ìñâ 0Ïó¥) ÏÇ∞Ï†êÎèÑ
    X = np.random.randn(100)
    Y = np.random.randn(100)
    plt.subplot(grid[0, 0]).scatter(X, Y)

    # (0Ìñâ 1~2Ïó¥) ÎßâÎåÄ Í∑∏ÎûòÌîÑ
    X = np.arange(10)
    Y = np.random.uniform(1, 10, 10)
    plt.subplot(grid[0, 1:]).bar(X, Y)

    # (1Ìñâ 0~1Ïó¥) ÏÑ† Í∑∏ÎûòÌîÑ
    X = np.linspace(0, 10, 100)
    Y = np.cos(X)
    plt.subplot(grid[1, :2]).plot(X, Y)

    # (1Ìñâ 2Ïó¥) 2D Ïù¥ÎØ∏ÏßÄ (ÏûÑÏùòÏùò ÎûúÎç§ Îç∞Ïù¥ÌÑ∞)
    Z = np.random.uniform(0, 1, (5, 5))
    plt.subplot(grid[1, 2]).imshow(Z)

    # tight_layout() ‚Üí Í∑∏ÎûòÌîÑ ÏöîÏÜå(ÎùºÎ≤®, Ï†úÎ™© Îì±)Í∞Ä Í≤πÏπòÏßÄ ÏïäÎèÑÎ°ù ÏûêÎèô Î∞∞Ïπò
    plt.tight_layout()
    plt.show()

# Îã§ÏñëÌïú wspace, hspace Ï°∞Ìï© ÏãúÏó∞
# (0.1,0.1) ‚Üí Í∞ÑÍ≤© Í±∞Ïùò ÏóÜÏùå
# (0.3,0.4) ‚Üí Í∏∞Î≥∏Ï†ÅÏù∏ Ï†ÅÎãπÌïú Ïó¨Î∞±
# (0.6,0.6) ‚Üí Í∞ÄÎ°ú¬∑ÏÑ∏Î°ú Í∞ÑÍ≤© Î™®Îëê ÎÑìÍ≤å
# (1.0,0.5) ‚Üí Í∞ÄÎ°ú Í∞ÑÍ≤©ÏùÄ ÌÅ¨Í≤å, ÏÑ∏Î°ú Í∞ÑÍ≤©ÏùÄ Ï†ÅÎãπÌûà
for ws, hs in [(0.1,0.1), (0.3,0.4), (0.6,0.6), (1.0,0.5)]:
    demo_wspace_hspace(ws, hs)
```

```python
# 3Î≤à Î¨∏Ï†ú ÎãµÏïà: GridSpecÏùÑ Ïù¥Ïö©Ìï¥ Ï≤´ Î≤àÏß∏ Ìñâ Ï†ÑÏ≤¥Î•º ÌïòÎÇòÏùò ÌÅ∞ Í∑∏ÎûòÌîÑÎ°ú Ìï©Ï≥êÎ≥¥Í∏∞
# ‚Üí 2Ìñâ 3Ïó¥ Íµ¨Ï°∞Ïùò Í≤©ÏûêÎ•º ÎßåÎì† Îí§, Ï≤´ Î≤àÏß∏ Ìñâ(0Ìñâ)ÏùÑ Í∞ÄÎ°úÎ°ú Î≥ëÌï©ÌïòÏó¨
#    ÌïòÎÇòÏùò ÌÅ∞ subplotÏúºÎ°ú ÏÇ¨Ïö©ÌïúÎã§.
# ‚Äª Ïù¥Î≤à Î¨∏Ï†úÏóêÏÑúÎäî "Ï≤´ Î≤àÏß∏ Ìñâ"Ïóê ÏßëÏ§ëÌïòÎØÄÎ°ú Î∂àÌïÑÏöîÌïú Îã§Î•∏ Í∑∏Î¶º(ÎßâÎåÄ, 2D Ïù¥ÎØ∏ÏßÄ Îì±)ÏùÄ Ï£ºÏÑù Ï≤òÎ¶¨

import matplotlib.pyplot as plt
import numpy as np

# GridSpecÏúºÎ°ú Ïó¨Î∞± Ï°∞Ï†ï Î∞è ÏòÅÏó≠ Î≥ëÌï©
grid = plt.GridSpec(2, 3, wspace=0.3, hspace=0.4)

# (0Ìñâ Ï†ÑÏ≤¥: 0Ïó¥~2Ïó¥) ‚Üí ÏÇ∞Ï†êÎèÑÎ•º ÌÅ∞ Í∑∏ÎûòÌîÑÎ°ú ÌëúÏãú
X = np.random.randn(100)
Y = np.random.randn(100)
plt.subplot(grid[0, :]).scatter(X, Y)   # [:]ÏùÄ "Î™®Îì† Ïó¥"ÏùÑ ÏùòÎØ∏ ‚Üí 0Ìñâ Ï†ÑÏ≤¥ Ï∞®ÏßÄ

# (0Ìñâ 1~2Ïó¥ Ìï©Ïπ®) ÎßâÎåÄ Í∑∏ÎûòÌîÑ ÏòàÏãúÎäî Ïù¥Î≤à Î¨∏Ï†úÏóêÏÑú Î∂àÌïÑÏöîÌïòÎØÄÎ°ú Ï£ºÏÑù Ï≤òÎ¶¨
# X = np.arange(10)
# Y = np.random.uniform(1, 10, 10)
# plt.subplot(grid[0, 1:]).bar(X, Y)

# (1Ìñâ 0~1Ïó¥) ‚Üí ÏÑ† Í∑∏ÎûòÌîÑ (cos Ìï®Ïàò)
X = np.linspace(0, 10, 100)
Y = np.cos(X)
plt.subplot(grid[1, :]).plot(X, Y)

# (1Ìñâ 2Ïó¥) ‚Üí 2D Ïù¥ÎØ∏ÏßÄ ÏòàÏãúÎäî Ïù¥Î≤à Î¨∏Ï†úÏóêÏÑú Î∂àÌïÑÏöîÌïòÎØÄÎ°ú Ï£ºÏÑù Ï≤òÎ¶¨
# Z = np.random.uniform(0, 1, (5, 5))
# plt.subplot(grid[1, 2]).imshow(Z)

plt.show()
```
